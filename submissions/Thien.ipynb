### Load the dataset

```python
df = pd.read_csv("data.csv")
print(df.shape)
df.head()
```

### Identify year columns

`year_cols_orig` contains the value variables (the years). It is dynamically generated by iterating through the DataFrame's columns and selecting those that contain a bracket `[`.

```python
year_cols_orig = [col for col in df.columns if '[' in col]
```

```python

print(year_cols_orig)
df[year_cols_orig].head()
```

### Handle non-standard missing values ('..' to NaN) and convert to float

```python
df[year_cols_orig] = df[year_cols_orig].replace('..', np.nan).astype(float)
```

Replace('..', np.nan): Finds all instances of '..' (which is the World Bank's specific placeholder for missing data) and replaces them with np.nan

### Clean column names to just the year integers for accurate time-series 
E.g., 2000 [YR2000] -> 2000

```python
year_mapping = {col: int(col.split(' ')[0]) for col in year_cols_orig}
df = df.rename(columns=year_mapping)
```

```python
print("Dataset Shape:", df.shape)
print("Time Horizon:", min(year_mapping.values()), "to", max(year_mapping.values()))
```

```python
df.head()
```

## 1.3 Handle Missing Values

### Missing Values Analysis

```python
df_ts = df[year_mapping.values()]

missing_per_year = df_ts.isna().sum()
missing_pct_per_year = (missing_per_year / len(df)) * 100
```

Extract only the time series portion (the year columns) for analysis and compute missing value count and percentage per Year.

```python
missing_summary = pd.DataFrame({
    'Missing_Count': missing_per_year,
    'Missing_Percentage (%)': missing_pct_per_year
})

print("\n--- Missing Values Per Year ---")
print(missing_summary)
```

Visualize missing patterns using a Seaborn heatmap.

```python
plt.figure(figsize=(14, 8))

# Dark Navy Blue for "Present" (False/0) and Bright Yellow for "Missing" (True/1)
custom_cmap = sns.color_palette(["#112A46", "#FCE029"])

# The y-axis is the row index (Country/Series), x-axis is the Year
sns.heatmap(df_ts.isna(), cbar=False, cmap=custom_cmap)

# Enhanced typography for the labels
plt.title('Missing Values Heatmap (Yellow = Missing, Dark Blue = Present)', fontsize=16, fontweight='bold', pad=15)
plt.ylabel('Row Index (Unique Country + Series)', fontsize=12)
plt.xlabel('Year Columns', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate x-labels slightly for better readability
plt.tight_layout()
plt.show()
```

Analysis of Missing Patterns:
- Horizontal gaps (Solid yellow lines stretching left to right) indicate specific Country/Series metrics are completely untracked.
- Vertical gaps (like solid yellow bands on specific years) indicate global dataset-wide missing years.
- Trailing missing values on the far right (2022-2024) show data that hasn't been collected or reported yet.

### Preprocessing Data

Convert all data in time-series table to float type

```python
df_ts = df_ts.astype(float)
```

Use a combination of 3 steps:
- Fill in the middle (interpolate).
- Backward fill for the beginning.
- Forward fill for the end.

```python
df_calc = df_ts.interpolate(method='values', axis=1) # Fill in the middle (interpolate)
df_calc = df_calc.bfill(axis=1)                      # Backward fill for the beginning
df_calc = df_calc.ffill(axis=1)                      # Forward fill for the end 
```

Apply to the original table to preserve data: Fill the missing values in the original table using the calculated table above .

```python
id_vars = ["Country Name", "Country Code", "Series Name", "Series Code"]
```

```python
df_time = df_ts.fillna(df_calc).astype(float)
df_final = pd.concat([df[id_vars], df_time], axis=1)
```

Analysis missing value after imputing.

```python
df_ts_final = df_final[year_mapping.values()]

missing_per_year = df_ts_final.isna().sum()
missing_pct_per_year = (missing_per_year / len(df)) * 100
```

```python
missing_summary = pd.DataFrame({
    'Missing_Count': missing_per_year,
    'Missing_Percentage (%)': missing_pct_per_year
})

print("\n--- Missing Values Per Year ---")
print(missing_summary)
```

## Analysis Objective 1: Comparing the Development Trends of High-Quality Human Resources (Tertiary Education) from 2000 to 2024

### 1. Indicator Used
* **Data field:** `School enrollment, tertiary (% gross)`
* **Meaning of the indicator:** This measures the ratio of total students enrolled in tertiary education (regardless of age) to the population of the official age group for that level. It reflects a country’s capacity for higher education, the level of advanced knowledge in society, and its focus on a knowledge-based economy.

### 2. Reason for Choosing This Indicator
* **Relevance to the objective:** To evaluate the internal strength and long-term competitiveness of major countries, tertiary education is a key measure. It shows the ability to supply skilled workers for high-tech industries and complex management systems.
* **Why this indicator:** Comparing this metric across the top 7 countries helps highlight shifts in global knowledge power, especially how emerging economies are catching up with traditional leaders.

### 3. Chart Type: **Line Chart with Markers**

### 4. Reason for Choosing This Chart
* **Nature of the data:** This is time-series data. A line chart is the most suitable way to show trends, growth rates, and key turning points over the 24-year period.

* **Handling real-world data:**
    * **Transparency:** Some countries (such as the United States, Japan, and Germany) have missing data in earlier years. The group keeps these gaps and only starts plotting from available data. Using **markers (dots)** helps clearly show actual data points and distinguish them from interpolated segments.
    * **Linear interpolation:** For short gaps within the timeline, linear interpolation is used to maintain a smooth trend instead of leaving flat lines or sudden jumps.

* **Comparison ability:** With 7 countries, using different colored lines makes it easier to compare their positions (who leads, who follows) and observe changes through the slope of each line.

```python
# 1. Load RAW data
df_raw = pd.read_csv('data.csv', na_values='..')
edu_indicator = "School enrollment, tertiary (% gross)"
df_edu_base = df_raw[df_raw['Series Name'] == edu_indicator].copy()
year_cols = [col for col in df_edu_base.columns if '[' in col]

# --- IMPORTANT STEP: CREATE TWO DATASETS ---

# Dataset 1: Interpolated data (used to draw continuous lines)
df_interp = df_edu_base.copy()
df_interp[year_cols] = df_interp[year_cols].interpolate(method='linear', axis=1, limit_area='inside')
df_interp_long = df_interp.melt(id_vars=['Country Name'], value_vars=year_cols, var_name='Year_Raw', value_name='Val')
df_interp_long['Year'] = df_interp_long['Year_Raw'].str.extract(r'(\d+)')[0].astype(int)

# Dataset 2: Original data (used for markers - only real values, missing ones remain NaN)
df_dots_long = df_edu_base.melt(id_vars=['Country Name'], value_vars=year_cols, var_name='Year_Raw', value_name='Val')
df_dots_long['Year'] = df_dots_long['Year_Raw'].str.extract(r'(\d+)')[0].astype(int)


plt.figure(figsize=(15, 8))
sns.set_style("whitegrid")

# Use a consistent color palette so both layers match
colors = sns.color_palette("tab10", n_colors=len(df_edu_base['Country Name'].unique()))
palette = dict(zip(df_edu_base['Country Name'].unique(), colors))

# LAYER 1: Draw LINES (using interpolated data, no markers)
sns.lineplot(
    data=df_interp_long, 
    x='Year', y='Val', hue='Country Name', 
    palette=palette, linewidth=2.5, alpha=0.6,
    marker=None, legend=True  # Keep legend here
)

# LAYER 2: Draw POINTS (using original data, missing years will have no points)
sns.scatterplot(
    data=df_dots_long, 
    x='Year', y='Val', hue='Country Name', 
    palette=palette, s=60, edgecolor='white',
    legend=False  # Disable legend here to avoid duplication
)

plt.title('Trends in Tertiary Education Enrollment of Major Countries (2000 - 2024)', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Tertiary Enrollment (% Gross)', fontsize=12)
plt.legend(title='Country', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.xticks(range(2000, 2025, 2), fontsize=10)
plt.xlim(2000, 2024)

plt.tight_layout()
plt.show()
```

### **ANALYSIS AND CONCLUSIONS (OBJECTIVE 1)**

#### **1. Key observations from the chart**
* **Data transparency:** The dots (markers) represent years with actual data from the World Bank.
  * **Continuous data group:** China, India, France, and the UK have fairly complete data since 2000, making their trends easy to follow.
  * **Missing data group:** The United States, Japan, and Germany lack data in earlier years (before 2013). The smooth lines without markers in this period are interpolated trends, used only to keep a general view, not exact values.

* **China’s rapid growth (orange line):** This is the most noticeable trend. Starting from the lowest level (below 10%), China shows very fast growth, especially after 2013. By 2024, it reaches about **78%**, surpassing France and Japan and joining the top group.

* **United States trend (blue line):** The US stays at a very high level (around 90%) from 2013 to 2020. However, after 2020, there is a **clear decline**, dropping to around 80% in 2024. This may reflect changes in its education system or economy.

* **European countries (Germany – red, UK – brown):** From 2013 onward, both countries show strong growth. By 2024, they reach around **80%**, confirming their strong position in high-quality education.

* **Japan’s stability (green line):** From 2013 to 2023, Japan remains almost flat at around 63–65%. This suggests the system may have reached a saturation point, possibly due to an aging population.

* **India (purple line):** India grows steadily but slowly. At around **35%** in 2024, it is still building its education base and remains far behind the other countries.

#### **2. Trends and insights**
* **Convergence trend:** The gap between major countries is shrinking. By 2024, the US, China, the UK, and Germany are all close to the same level (around 80%).

* **Shift in knowledge power:** The center of higher education growth is moving toward East Asia (especially China), while Western countries show signs of slowing down or restructuring.

#### **3. Conclusion**
The chart shows clearly that **China** is rapidly strengthening its human resources through strong investment in education. The **United States** remains strong but is showing signs of slowing down, while **Japan** appears to have reached its limit. For **India**, higher education still has a lot of room to grow and will be important for its future development.

## Analysis Objective 2: Comparing Population Structure and Aging Levels of Major Countries (2024)

### 1. Indicators Used
* **Data fields:**  
    * `Population ages 0-14 (% of total population)` – Children  
    * `Population ages 15-64 (% of total population)` – Working-age population  
    * `Population ages 65 and above (% of total population)` – Elderly  
* **Meaning of the indicators:** These indicators describe the population structure of a country. Together, they sum to 100%, helping us see whether a country is in a “golden population” phase or facing population aging.

### 2. Reason for Choosing These Indicators
* **Relevance to the objective:** Age structure directly affects a country’s long-term economic and social strength. The 15–64 group is the main workforce, while the 0–14 and 65+ groups represent dependency in terms of education and social support.
* **Why these indicators:** Comparing these three groups across countries helps us predict future labor supply (like India) or pressure on healthcare and pension systems (like Japan and Germany).

### 3. Chart Type: **Stacked Bar Chart**

### 4. Reason for Choosing This Chart
* **Nature of the data:** Since all groups add up to 100%, a stacked bar chart is the best way to show how each part contributes to the whole.
* **Comparison ability:** It allows direct comparison of age group sizes across all 7 countries on the same axis.
* **Clarity:** Clear color separation helps quickly identify which country is “younger” and which is more “aged.”

```python
# 1. Load data from the preprocessed file
df_imputed = pd.read_csv('data_imputed.csv')

# 2. List of required indicators
indicators = [
    'Population ages 0-14 (% of total population)',
    'Population ages 15-64 (% of total population)',
    'Population ages 65 and above (% of total population)'
]

# 3. Filter data for the year 2024 and selected indicators
df_2024 = df_imputed[df_imputed['Series Name'].isin(indicators)].copy()

# Convert to pivot table format for easier plotting
# Rows: Country, Columns: Indicators, Values: 2024 data
df_plot = df_2024.pivot(index='Country Name', columns='Series Name', values='2024')

# Rename columns to make them shorter and cleaner
df_plot.columns = ['Young (0-14)', 'Working (15-64)', 'Old (65+)']

# Sort countries by elderly population (Old) in descending order for better visualization
df_plot = df_plot.sort_values('Old (65+)', ascending=False)

# 4. Plot stacked bar chart
plt.figure(figsize=(14, 8))

# Define colors
colors = ['#377eb8', '#ff7f00', '#984ea3']  # Blue (Young), Green (Working), Red (Old)

# Plot each layer
p1 = plt.bar(df_plot.index, df_plot['Young (0-14)'], color=colors[0], label='Children (0-14)')
p2 = plt.bar(df_plot.index, df_plot['Working (15-64)'], 
             bottom=df_plot['Young (0-14)'], 
             color=colors[1], label='Working age (15-64)')
p3 = plt.bar(df_plot.index, df_plot['Old (65+)'], 
             bottom=df_plot['Young (0-14)'] + df_plot['Working (15-64)'], 
             color=colors[2], label='Elderly (65+)')

# 5. Add percentage labels inside the bars
for i in range(len(df_plot)):
    # Label for elderly (top section)
    plt.text(i, 100 - df_plot['Old (65+)'].iloc[i]/2, f"{df_plot['Old (65+)'].iloc[i]:.1f}%", 
             ha='center', va='center', fontweight='bold', fontsize=10)
    
    # Label for children (bottom section)
    plt.text(i, df_plot['Young (0-14)'].iloc[i]/2, f"{df_plot['Young (0-14)'].iloc[i]:.1f}%", 
             ha='center', va='center', fontweight='bold', fontsize=10)

# 6. Customize the chart
plt.title('Population Structure of Major Countries (2024)', fontsize=16, fontweight='bold', pad=20)
plt.ylabel('Percentage of Population (%)', fontsize=12)
plt.legend(title='Age Groups', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.ylim(0, 110)  # Add space for labels

plt.tight_layout()
plt.show()
```

### **ANALYSIS AND CONCLUSIONS (OBJECTIVE 2)**

#### **1. Key observations from the chart (2024)**
Based on the percentages of the three age groups shown in the chart, we can observe:

* **Elderly population (65+):**  
    * **Japan** has the highest share at **29.8%**.  
    * It is followed by European countries: **Germany (23.2%)** and **France (22.1%)**.  
    * The **United States** and the **United Kingdom** are lower, at **17.9%** and **19.5%**.  
    * **India** has the lowest share, only **7.1%**.

* **Children (0–14):**  
    * **India** clearly leads with **24.6%**.  
    * **Japan** has the lowest share at **11.4%**.  
    * Other countries (US, UK, France, Germany, China) range from **13.9% to 17.3%**.

* **Working-age population (15–64):**  
    * The middle segment (orange in the chart) is the largest in all countries (always above 50%).  
    * **China** and **India** visually have a larger working-age share compared to Japan.

#### **2. Trends and insights**
* **Strong contrast between extremes:** Japan’s elderly share is about **2.6 times** its child population (29.8% vs 11.4%). In contrast, India’s child population is about **3.4 times** its elderly population (24.6% vs 7.1%).

* **Balanced group:** The **United States** and the **United Kingdom** have a more balanced structure, where the gap between children and elderly is relatively small (less than 2.5%).

* **China’s position:** China has a child share (**16.0%**) and elderly share (**14.7%**) that are quite close. This suggests it is moving toward a more balanced structure, similar to developed countries like the US.

#### **3. Conclusion**
The 2024 data shows a clear difference in population structure among the 7 countries:

1. **Aging societies:** Japan, Germany, and France, where the elderly population is clearly larger than the child population.  
2. **Young society:** India, with a much higher share of children, showing strong future labor potential.  
3. **Transitional group:** China and the remaining countries are in a middle stage, where the shares of dependent groups (children and elderly) are becoming closer.

## Analysis Objective 3: The Relationship between “Wealth” (Economy) and “Urbanization” (Society) – 2024

### 1. Indicators Used
* **Data fields:**  
    * `GDP per capita (current US$)` – GDP per capita (represents economic strength)  
    * `Urban population (% of total population)` – Urban population share (represents social transformation)  

* **Meaning of the indicators:** Combining these two indicators helps us understand the relationship between economic wealth and modern living. It answers a simple question: *Do richer countries tend to have more people living in urban areas?*

### 2. Reason for Choosing These Indicators
* **Relevance to the objective:** This is a multi-dimensional analysis to explore the link between economic and social development.

* **Why these indicators:** Comparing them helps group countries into different stages:  
    - Highly developed with high income and high urbanization  
    - Rapidly developing and urbanizing  
    - Still having room to grow  

### 3. Chart Type: **Scatter Plot with Trend Line**

### 4. Reason for Choosing This Chart
* **Nature of the data:** A scatter plot is the best way to show the relationship (correlation) between two numerical variables.

* **Clarity:** Adding a trend line (regression line) makes it easier to see the direction of the relationship (positive or negative) and how closely countries follow the overall trend.

```python
# 1. Load data
df_imputed = pd.read_csv('data_imputed.csv')

# 2. Filter the two required indicators for the year 2024
indicators = [
    'GDP per capita (current US$)',
    'Urban population (% of total population)'
]
df_2024 = df_imputed[df_imputed['Series Name'].isin(indicators)].copy()

# 3. Pivot the data so each indicator becomes a column
df_plot = df_2024.pivot(index='Country Name', columns='Series Name', values='2024').reset_index()
df_plot.columns = ['Country', 'GDP_per_capita', 'Urban_Rate']

# 4. Plot the chart
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")

# Use a colorblind-friendly palette
palette = sns.color_palette("colorblind", n_colors=len(df_plot))

# Plot data points
scatter = sns.scatterplot(
    data=df_plot, 
    x='GDP_per_capita', 
    y='Urban_Rate', 
    hue='Country', 
    s=300, 
    palette=palette,
    edgecolor='black',
    alpha=0.9
)

# Add trend line (avoid red/green conflict)
sns.regplot(
    data=df_plot, 
    x='GDP_per_capita', 
    y='Urban_Rate', 
    scatter=False, 
    color='gray', 
    line_kws={"ls":"--", "alpha":0.4}
)

# 5. Add country labels next to each point for clarity
for i in range(df_plot.shape[0]):
    plt.text(
        x=df_plot.GDP_per_capita.iloc[i] + 1000, 
        y=df_plot.Urban_Rate.iloc[i] + 0.5, 
        s=df_plot.Country.iloc[i], 
        fontsize=11, 
        fontweight='bold'
    )

# 6. Customize the chart
plt.title('Relationship between GDP per Capita and Urbanization (2024)', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('GDP per capita (current US$)', fontsize=12)
plt.ylabel('Urban population (%)', fontsize=12)

# Format X-axis with thousands separators
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))

plt.legend(title='Country', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### **ANALYSIS AND CONCLUSIONS (OBJECTIVE 3)**

#### **1. Key observations from the chart (2024)**
Based on the positions of the 7 countries on the chart (GDP per capita vs. Urban population), we can see:

* **Income (X-axis – GDP per capita):**

    * **India** is the farthest to the left, with the lowest income.

* **Relationship between the two indicators:**
    * The dashed trend line goes upward from left to right, showing a **positive relationship**: countries with higher income tend to have higher urbanization.

#### **2. Trends and insights**
* **Clear grouping of development levels:**  
The chart divides the countries into three groups:
    1. **Highly urbanized group:** Japan, UK, US, Germany, and France all have urbanization rates above 80%. Japan stands out with very high urbanization despite not having the highest GDP.

## Analysis Objective 4: Internet Adoption Growth and the Rise of the Digital Society (2000–2024)

### 1. Indicator Used

* **Meaning of the indicator:**  
This measures the percentage of people in a country who use the Internet. It is a basic indicator to evaluate the level of digitalization, access to online knowledge, and the development of technology infrastructure.
### 2. Reason for Choosing This Indicator
* **Relevance to the objective:** In the 21st century, a country’s strength is not only about physical resources but also digital connectivity.
* **Why this indicator:** Tracking this metric from 2000 to 2024 helps compare how fast different countries are changing. It shows which countries led the digital transformation and which ones are catching up quickly.
### 3. Chart Type: **Heatmap**
### 4. Reason for Choosing This Chart
* **Nature of the data:** A heatmap is suitable for showing three dimensions at once (Country × Time × Percentage) in a clear way.
* **Clarity:** It allows easy comparison both across time (each country’s progress) and across countries at the same time.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load data
df_imputed = pd.read_csv('data_imputed.csv')

# 2. Filter Internet usage indicator
internet_indicator = "Individuals using the Internet (% of population)"
df_internet = df_imputed[df_imputed['Series Name'] == internet_indicator].copy()

# 3. Select specific years to avoid overcrowding the chart (e.g., every 4 years + latest year)
years_to_show = ['2000', '2004', '2008', '2012', '2016', '2020', '2024']
# Keep only years that exist in the dataset
years_to_show = [y for y in years_to_show if y in df_internet.columns]

# 4. Create pivot table for Heatmap
heatmap_data = df_internet.set_index('Country Name')[years_to_show]

# Sort countries by 2024 values (descending) for better visualization
heatmap_data = heatmap_data.sort_values(by='2024', ascending=False)

# 5. Plot Heatmap
plt.figure(figsize=(14, 10))

# Use colorblind-friendly colormap
sns.heatmap(
    heatmap_data, 
    annot=True, 
    fmt=".1f", 
    cmap="YlGnBu", 
    linewidths=.8, 
    cbar_kws={'label': 'Internet users (%)'},
    annot_kws={"size": 11, "weight": "bold"}
)

# 6. Customize chart
plt.title('Heatmap: Global Internet Adoption (2000 - 2024)', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Country', fontsize=12)

plt.tight_layout()
plt.show()
```

### **ANALYSIS AND CONCLUSIONS (OBJECTIVE 4)**

#### **1. Key observations from the chart (2000–2024)**
Based on the values and the color scale (from light yellow = low to dark blue = high), we can observe:

* **Overall change:** There is a clear shift from light colors in 2000 to dark blue in 2024 across all countries. This shows strong global growth in Internet usage over the past 24 years.

* **Top countries in 2024 (above 90%):**  
    * The **United Kingdom** has the highest rate at **96.3%**.  
    * **Germany** follows with **93.5%**.  
    * The **United States** is close behind at **93.1%**.  
    * **China** has also reached a high level at **92.0%**.

* **China’s rapid growth:** China shows the fastest change. From only **1.8%** in 2000 (lowest level), it increased steadily every few years (7.3% → 22.6% → 42.3% → 53.2% → 70.1%) and reached **92.0%** in 2024, joining the top group.

* **India’s position:** India started at the lowest level (**0.5%** in 2000) and remained low for many years. However, it jumped from **43.4%** in 2020 to **55.9%** in 2024. It is the only country still in the mid-range (not yet at the highest level).

* **Stability and saturation:**  
Countries like Japan and France have stayed above **80%** since around 2012.  
Japan is especially stable, staying around **87%–93%** in the last 8 years.

#### **2. Trends and insights**
* **Convergence in digital infrastructure:** The gap between developed countries and China has almost disappeared by 2024 (the difference between the UK and China is only about 4.3%).

* **Different growth patterns:**  
    * Western countries (US, UK, Germany) show steady growth over time.  
    * China shows very fast growth, especially from 2008 to 2024.  
    * India is in an early rapid growth phase, with the biggest increase in recent years.

#### **3. Conclusion**
The heatmap shows that a **digital society** has formed in most major countries by 2024.  
The **United Kingdom, Germany, and the United States** remain leading countries.  

However, **China’s rapid rise** to the same level as developed countries shows strong progress in digital infrastructure.  
**India**, although still behind, has strong potential for digital growth based on its recent rapid increase.
