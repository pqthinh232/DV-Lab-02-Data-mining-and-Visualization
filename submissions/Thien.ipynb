### Load the dataset

```python
df = pd.read_csv("..\data.csv")
print(df.shape)
df.head()
```

### Identify year columns

`year_cols_orig` contains the value variables (the years). It is dynamically generated by iterating through the DataFrame's columns and selecting those that contain a bracket `[`.

```python
year_cols_orig = [col for col in df.columns if '[' in col]
```

```python

print(year_cols_orig)
df[year_cols_orig].head()
```

### Handle non-standard missing values ('..' to NaN) and convert to float

```python
df[year_cols_orig] = df[year_cols_orig].replace('..', np.nan).astype(float)
```

Replace('..', np.nan): Finds all instances of '..' (which is the World Bank's specific placeholder for missing data) and replaces them with np.nan

### Clean column names to just the year integers for accurate time-series 
E.g., 2000 [YR2000] -> 2000

```python
year_mapping = {col: int(col.split(' ')[0]) for col in year_cols_orig}
df = df.rename(columns=year_mapping)
```

```python
print("Dataset Shape:", df.shape)
print("Time Horizon:", min(year_mapping.values()), "to", max(year_mapping.values()))
```

```python
df.head()
```

## 1.3 Handle Missing Values

### Missing Values Analysis

```python
df_ts = df[year_mapping.values()]

missing_per_year = df_ts.isna().sum()
missing_pct_per_year = (missing_per_year / len(df)) * 100
```

Extract only the time series portion (the year columns) for analysis and compute missing value count and percentage per Year.

```python
missing_summary = pd.DataFrame({
    'Missing_Count': missing_per_year,
    'Missing_Percentage (%)': missing_pct_per_year
})

print("\n--- Missing Values Per Year ---")
print(missing_summary)
```

Visualize missing patterns using a Seaborn heatmap.

```python
plt.figure(figsize=(14, 8))

# Dark Navy Blue for "Present" (False/0) and Bright Yellow for "Missing" (True/1)
custom_cmap = sns.color_palette(["#112A46", "#FCE029"])

# The y-axis is the row index (Country/Series), x-axis is the Year
sns.heatmap(df_ts.isna(), cbar=False, cmap=custom_cmap)

# Enhanced typography for the labels
plt.title('Missing Values Heatmap (Yellow = Missing, Dark Blue = Present)', fontsize=16, fontweight='bold', pad=15)
plt.ylabel('Row Index (Unique Country + Series)', fontsize=12)
plt.xlabel('Year Columns', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate x-labels slightly for better readability
plt.tight_layout()
plt.show()
```

Analysis of Missing Patterns:
- Horizontal gaps (Solid yellow lines stretching left to right) indicate specific Country/Series metrics are completely untracked.
- Vertical gaps (like solid yellow bands on specific years) indicate global dataset-wide missing years.
- Trailing missing values on the far right (2022-2024) show data that hasn't been collected or reported yet.

### Preprocessing Data

Convert all data in time-series table to float type

```python
df_ts = df_ts.astype(float)
```

Use a combination of 3 steps:
- Fill in the middle (interpolate).
- Backward fill for the beginning.
- Forward fill for the end.

```python
df_calc = df_ts.interpolate(method='values', axis=1) # Fill in the middle (interpolate)
df_calc = df_calc.bfill(axis=1)                      # Backward fill for the beginning
df_calc = df_calc.ffill(axis=1)                      # Forward fill for the end 
```

Apply to the original table to preserve data: Fill the missing values in the original table using the calculated table above .

```python
id_vars = ["Country Name", "Country Code", "Series Name", "Series Code"]
```

```python
df_time = df_ts.fillna(df_calc).astype(float)
df_final = pd.concat([df[id_vars], df_time], axis=1)
```

Analysis missing value after imputing.

```python
df_ts_final = df_final[year_mapping.values()]

missing_per_year = df_ts_final.isna().sum()
missing_pct_per_year = (missing_per_year / len(df)) * 100
```

```python
missing_summary = pd.DataFrame({
    'Missing_Count': missing_per_year,
    'Missing_Percentage (%)': missing_pct_per_year
})

print("\n--- Missing Values Per Year ---")
print(missing_summary)
```

## Analysis Objective 5: Health Quality and Life Outcomes (2000 – 2023)

### 1. Indicators Used
*   **Data fields:**  
    *   `Life expectancy at birth, total (years)` – Overall health outcome.
    *   `Mortality rate, infant (per 1,000 live births)` – Healthcare quality proxy.
*   **Meaning of the indicators:** These reflect the efficiency of the national health system. Life expectancy shows the long-term result, while infant mortality is sensitive to immediate care and nutrition.

### 2. Reason for Choosing These Indicators
*   **Relevance to the objective:** They represent the "biological success" of a society. Higher wealth should logically lead to longer, healthier lives.
*   **Why these indicators:** To track the "catch-up" progress of emerging nations (India/China) versus the stability or stagnation in developed ones.

### 3. Chart Type: **Multi-Line Chart**

### 4. Reason for Choosing This Chart
*   **Nature of the data:** Best for showing continuous trends over 24 years.
*   **Comparison ability:** Shows which countries are improving faster and where global crises (like 2020) caused dips.




```python
# 1. Load data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_imputed = pd.read_csv('data_imputed.csv')
countries = ['United States', 'China', 'Japan', 'Germany', 'India', 'United Kingdom', 'France']
years = [str(y) for y in range(2000, 2024)]

# 2. Filter data
indicator = 'Life expectancy at birth, total (years)'
df_obj5 = df_imputed[(df_imputed['Country Name'].isin(countries)) & (df_imputed['Series Name'] == indicator)]
df_plot = df_obj5.set_index('Country Name')[years].T
df_plot.index = df_plot.index.astype(int)

# 3. Visualization
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_plot, dashes=False, palette='tab10', linewidth=2.5)
plt.title('Analysis Objective 5: Life Expectancy Trends (2000-2023)', fontsize=14, fontweight='bold')
plt.ylabel('Years of Life')
plt.xlabel('Year')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```



### **ANALYSIS AND CONCLUSIONS (OBJECTIVE 5)**
**Objective 5: Longitudinal Trends in Life Expectancy (2000-2023)**

**1. Key observations from the chart**
*   **Universal Upward Trajectory:**
    *   From 2000 to 2019, all seven nations showed a steady, near-linear increase.
    *   **India** exhibited the most aggressive growth of all countries studied.
    *   Data shows India rising from **~62.5 years to 70.9 years** pre-pandemic.
*   **The Elite Longevity Group (Japan, France, Germany, UK):**
    *   These nations form a tight, "high-performance" health cluster.
    *   **Japan** leads consistently, reaching a peak of nearly **84.5 years**.
    *   European models (France/UK) maintain stability above **80 years**.
*   **The 2020-2021 Volatility:**
    *   The pandemic caused a significant "V-shaped" disruption globally.
    *   The **US** saw a sharp decline from **~79 to ~76.4 years** (a 20-year setback).
    *   **India** experienced a sharp, temporary collapse in its growth trend.
    *   **China** remained remarkably stable, continuing its climb toward **78 years**.

**2. Trends and insights**
*   **Converging Health Outcomes:**
    *   The gap between Japan (highest) and India (lowest) was **~18 years** in 2000.
    *   By 2023, this gap narrowed significantly due to globalized medical tech.
*   **Systemic Resilience vs. Wealth:**
    *   Data reveals that high GDP does not guarantee health resilience.
    *   The **US** proved more vulnerable to shocks than **Japan** or **Germany**.
    *   Public health equity appears more vital for stability than total spending.

**3. Conclusion**
Global health is improving, but the pace is uneven. **China and India** are rapidly catching up, while the **United States** faces a unique crisis of stability compared to G7 peers.





## Analysis Objective 6: Labor Market Resilience (2019 vs 2022)

### 1. Indicators Used
*   **Data fields:**  
    *   `Unemployment, total (% of total labor force) (modeled ILO estimate)`
*   **Meaning of the indicators:** Measures the percentage of the workforce unable to find a job.

### 2. Reason for Choosing These Indicators
*   **Relevance to the objective:** Resilience is measured by how labor markets recovered from the COVID-19 shock. 
*   **Why these indicators:** 2019 represents the pre-pandemic baseline, and 2022 represents the post-crisis recovery status.

### 3. Chart Type: **Grouped Bar Chart**

### 4. Reason for Choosing This Chart
*   **Nature of the data:** Allows side-by-side comparison of pre-crisis and post-crisis metrics.
*   **Clarity:** Clearly shows which countries returned to normal and which still struggle with high unemployment.






```python
# 1. Filter for Objective 6
indicator_unemp = 'Unemployment, total (% of total labor force) (modeled ILO estimate)'
df_obj6 = df_imputed[(df_imputed['Country Name'].isin(countries)) & (df_imputed['Series Name'] == indicator_unemp)]

# 2. Reshape data for plotting
df_bar = df_obj6[['Country Name', '2019', '2022']].melt(id_vars='Country Name', var_name='Year', value_name='Rate')

# 3. Visualization
plt.figure(figsize=(12, 6))
sns.barplot(data=df_bar, x='Country Name', y='Rate', hue='Year', palette='viridis')
plt.title('Analysis Objective 6: Unemployment Rate Comparison (2019 vs 2022)', fontsize=14, fontweight='bold')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```


### **ANALYSIS AND CONCLUSIONS (OBJECTIVE 6)**
**Objective 6: Labor Market Resilience Post-Pandemic (2019 vs. 2022)**

**1. Key observations from the chart**
*   **The French Structural Success:**
    *   **France** recorded the most notable improvement in this period.
    *   Historically high unemployment (**~8.4%** in 2019) was lowered to **~7.3%**.
    *   This represents the lowest rate in France’s recent economic history.
*   **Low-Volatility Anchors (Japan & Germany):**
    *   **Japan** remains the gold standard with a rate steady at **~2.4% - 2.6%**.
    *   **Germany** follows a "labor hoarding" model, holding steady at **~3.1%**.
    *   Both nations prioritize job security over market fluctuations.
*   **India’s Statistical Recovery:**
    *   **India** saw a significant drop from **~6.5% to ~4.8%**.
    *   This often indicates workers moving back into the informal sector.
    *   It lacks the high-quality industrial job creation seen in Western peers.

**2. Trends and insights**
*   **Market Elasticity:**
    *   The **US** and **UK** exhibit very high labor market "elasticity."
    *   They shed millions of jobs in 2020 but re-absorbed them by 2022.
    *   Both returned to a pre-pandemic baseline of **~3.6%**.
*   **Policy Effectiveness:**
    *   Stable rates by 2022 prove that global stimulus packages worked.
    *   Large-scale interventions prevented a long-term global depression.

**3. Conclusion**
The global labor market has shifted from crisis to "tightness." **France** is solving long-term issues, while the **US and UK** prove that flexible markets recover fastest.




## Analysis Objective 7: Gender Equality in the Workforce (2022)

### 1. Indicators Used
*   **Data fields:**  
    *   `Labor force participation rate, female (% of female population ages 15+)`
    *   `Labor force participation rate, male (% of male population ages 15+)`
*   **Meaning of the indicators:** Tracks the gap between men and women active in the economy.

### 2. Reason for Choosing These Indicators
*   **Relevance to the objective:** A narrow gap represents high social equality and economic modernization.
*   **Why these indicators:** This reveals cultural barriers in labor participation across different continents.

### 3. Chart Type: **Lollipop Chart**

### 4. Reason for Choosing This Chart
*   **Clarity:** It visualizes the "distance" (the gender gap) for each country on a single axis, making the social gap intuitive to understand.




```python
# 1. Prepare data
f_ind = 'Labor force participation rate, female (% of female population ages 15+) (modeled ILO estimate)'
m_ind = 'Labor force participation rate, male (% of male population ages 15+) (modeled ILO estimate)'

df_f = df_imputed[(df_imputed['Country Name'].isin(countries)) & (df_imputed['Series Name'] == f_ind)][['Country Name', '2022']].rename(columns={'2022': 'Female'})
df_m = df_imputed[(df_imputed['Country Name'].isin(countries)) & (df_imputed['Series Name'] == m_ind)][['Country Name', '2022']].rename(columns={'2022': 'Male'})
df_gap = pd.merge(df_f, df_m, on='Country Name').sort_values('Male')

# 2. Visualization
plt.figure(figsize=(10, 6))
plt.hlines(y=df_gap['Country Name'], xmin=df_gap['Female'], xmax=df_gap['Male'], color='grey', alpha=0.6)
plt.scatter(df_gap['Female'], df_gap['Country Name'], color='#e74c3c', label='Female', s=100, zorder=3)
plt.scatter(df_gap['Male'], df_gap['Country Name'], color='#3498db', label='Male', s=100, zorder=3)
plt.title('Analysis Objective 7: Gender Gap in Labor Participation (2022)', fontsize=14, fontweight='bold')
plt.xlabel('Participation Rate (%)')
plt.legend()
plt.show()
```


### **ANALYSIS AND CONCLUSIONS (OBJECTIVE 7)**
**Objective 7: The Economic Gender Gap and Workforce Inclusivity (2022)**

**1. Key observations from the chart**
*   **India’s Extreme Disparity:**
    *   **India** is the most significant outlier in the entire study.
    *   Male participation is high (**~76%**), but female participation is only **~28%**.
    *   This **47-point gap** is the largest within the G20 countries.
*   **China’s Industrial Integration:**
    *   **China** exhibits a unique profile with female participation at **~61%**.
    *   This is higher than the **US (~56%)** and **France (~52%)**.
    *   Reflects a legacy of integrating women into manufacturing and services.
*   **The European Parity Group:**
    *   **The UK, Germany, and France** show the narrowest gender gaps.
    *   Most gaps are under **10-12%** in these nations.
    *   Indicates mature institutional support for working families.

**2. Trends and insights**
*   **The Opportunity Cost of Inequality:**
    *   India’s data suggests massive "untapped" economic potential.
    *   Closing the gap could boost India's GDP by trillions over the next decade.
*   **Modernization Correlation:**
    *   There is a direct link between service-based economies and gender parity.
    *   Knowledge Economies (UK/US) require diverse talent to sustain growth.

**3. Conclusion**
Gender equality remains a tale of two worlds. The **West and China** have integrated women into the economy, while **India** faces structural barriers that limit its output.






## Analysis Objective 8: Health Efficiency (Spending vs. Outcome) - 2020

### 1. Indicators Used
*   **Data fields:**  
    *   `Current health expenditure per capita (current US$)` – The Cost.
    *   `Life expectancy at birth, total (years)` – The Result.
*   **Meaning of the indicators:** This determines the "Value for Money" in healthcare.

### 2. Reason for Choosing These Indicators
*   **Relevance to the objective:** It tests whether spending more money leads to longer lives.
*   **Why these indicators:** Reveals the efficiency of the US private-led system compared to the Japanese/European public models.

### 3. Chart Type: **Scatter Plot with Trend Line**

### 4. Reason for Choosing This Chart
*   **Nature of the data:** Best way to visualize the correlation between economic input and social outcome.
*   **Clarity:** Clearly identifies "outliers"—countries that spend heavily but achieve poor results.





```python
# 1. Prepare data for 2020
exp_ind = 'Current health expenditure per capita (current US$)'
life_ind = 'Life expectancy at birth, total (years)'

df_exp = df_imputed[(df_imputed['Country Name'].isin(countries)) & (df_imputed['Series Name'] == exp_ind)][['Country Name', '2020']].rename(columns={'2020': 'Spending'})
df_life = df_imputed[(df_imputed['Country Name'].isin(countries)) & (df_imputed['Series Name'] == life_ind)][['Country Name', '2020']].rename(columns={'2020': 'LifeExp'})
df_scatter = pd.merge(df_exp, df_life, on='Country Name')

# 2. Visualization
plt.figure(figsize=(10, 7))
sns.regplot(data=df_scatter, x='Spending', y='LifeExp', scatter_kws={'s':200, 'alpha':0.7}, line_kws={"color": "red", "ls": "--"})

for i in range(df_scatter.shape[0]):
    plt.text(df_scatter.Spending[i]+150, df_scatter.LifeExp[i], df_scatter['Country Name'][i], fontsize=10, fontweight='bold')

plt.title('Analysis Objective 8: Health Efficiency (Spending vs Life Exp) - 2020', fontsize=14, fontweight='bold')
plt.xlabel('Health Spending per Capita (USD)')
plt.ylabel('Life Expectancy (Years)')
plt.show()
```

### **ANALYSIS AND CONCLUSIONS (OBJECTIVE 8)**
**Objective 8: Healthcare Efficiency – Spending vs. Outcome (2020)**

**1. Key observations from the chart**
*   **The US Efficiency Crisis:**
    *   The **US** is a massive outlier on the extreme right of the X-axis.
    *   Spending exceeds **$11,000 per capita** annually.
    *   Longevity (**~77 years**) is lower than **China’s**, which spends far less.
*   **The Efficiency Frontier (Japan & France):**
    *   **Japan and France** occupy the "Optimal Zone" of the chart.
    *   Highest life outcomes (**83-85 years**) with moderate spending.
    *   Expenditures range between **$4,500 and $5,500**.
*   **The High-Gain Zone (India & China):**
    *   Both nations are positioned in the bottom-left corner.
    *   At this stage, every additional dollar leads to a vertical jump in life expectancy.

**2. Trends and insights**
*   **Diminishing Marginal Returns:**
    *   The trend line shows that after **~$5,000**, more money doesn't buy more life.
    *   System design (Public vs. Private) becomes the deciding factor.
*   **The Public Health Advantage:**
    *   Public-centric systems (**UK, Japan, France**) significantly outperform.
    *   They deliver better "value for money" than the private-heavy **US** model.

**3. Conclusion**
Healthcare success is a matter of efficiency, not just dollars. The **US** is the most expensive and least efficient, while **Japan** achieves the best outcomes at moderate cost.
