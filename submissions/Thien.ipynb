### Load the dataset

```python
df = pd.read_csv("data.csv")
print(df.shape)
df.head()
```

### Identify year columns

`year_cols_orig` contains the value variables (the years). It is dynamically generated by iterating through the DataFrame's columns and selecting those that contain a bracket `[`.

```python
year_cols_orig = [col for col in df.columns if '[' in col]
```

```python

print(year_cols_orig)
df[year_cols_orig].head()
```

### Handle non-standard missing values ('..' to NaN) and convert to float

```python
df[year_cols_orig] = df[year_cols_orig].replace('..', np.nan).astype(float)
```

Replace('..', np.nan): Finds all instances of '..' (which is the World Bank's specific placeholder for missing data) and replaces them with np.nan

### Clean column names to just the year integers for accurate time-series 
E.g., 2000 [YR2000] -> 2000

```python
year_mapping = {col: int(col.split(' ')[0]) for col in year_cols_orig}
df = df.rename(columns=year_mapping)
```

```python
print("Dataset Shape:", df.shape)
print("Time Horizon:", min(year_mapping.values()), "to", max(year_mapping.values()))
```

```python
df.head()
```

## 1.3 Handle Missing Values

### Missing Values Analysis

```python
df_ts = df[year_mapping.values()]

missing_per_year = df_ts.isna().sum()
missing_pct_per_year = (missing_per_year / len(df)) * 100
```

Extract only the time series portion (the year columns) for analysis and compute missing value count and percentage per Year.

```python
missing_summary = pd.DataFrame({
    'Missing_Count': missing_per_year,
    'Missing_Percentage (%)': missing_pct_per_year
})

print("\n--- Missing Values Per Year ---")
print(missing_summary)
```

Visualize missing patterns using a Seaborn heatmap.

```python
plt.figure(figsize=(14, 8))

# Dark Navy Blue for "Present" (False/0) and Bright Yellow for "Missing" (True/1)
custom_cmap = sns.color_palette(["#112A46", "#FCE029"])

# The y-axis is the row index (Country/Series), x-axis is the Year
sns.heatmap(df_ts.isna(), cbar=False, cmap=custom_cmap)

# Enhanced typography for the labels
plt.title('Missing Values Heatmap (Yellow = Missing, Dark Blue = Present)', fontsize=16, fontweight='bold', pad=15)
plt.ylabel('Row Index (Unique Country + Series)', fontsize=12)
plt.xlabel('Year Columns', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate x-labels slightly for better readability
plt.tight_layout()
plt.show()
```

Analysis of Missing Patterns:
- Horizontal gaps (Solid yellow lines stretching left to right) indicate specific Country/Series metrics are completely untracked.
- Vertical gaps (like solid yellow bands on specific years) indicate global dataset-wide missing years.
- Trailing missing values on the far right (2022-2024) show data that hasn't been collected or reported yet.

### Preprocessing Data

Convert all data in time-series table to float type

```python
df_ts = df_ts.astype(float)
```

Use a combination of 3 steps:
- Fill in the middle (interpolate).
- Backward fill for the beginning.
- Forward fill for the end.

```python
df_calc = df_ts.interpolate(method='values', axis=1) # Fill in the middle (interpolate)
df_calc = df_calc.bfill(axis=1)                      # Backward fill for the beginning
df_calc = df_calc.ffill(axis=1)                      # Forward fill for the end 
```

Apply to the original table to preserve data: Fill the missing values in the original table using the calculated table above .

```python
id_vars = ["Country Name", "Country Code", "Series Name", "Series Code"]
```

```python
df_time = df_ts.fillna(df_calc).astype(float)
df_final = pd.concat([df[id_vars], df_time], axis=1)
```

Analysis missing value after imputing.

```python
df_ts_final = df_final[year_mapping.values()]

missing_per_year = df_ts_final.isna().sum()
missing_pct_per_year = (missing_per_year / len(df)) * 100
```

```python
missing_summary = pd.DataFrame({
    'Missing_Count': missing_per_year,
    'Missing_Percentage (%)': missing_pct_per_year
})

print("\n--- Missing Values Per Year ---")
print(missing_summary)
```
