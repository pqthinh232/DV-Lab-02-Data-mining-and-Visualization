### Load the dataset

```python
df = pd.read_csv("data.csv")
print(df.shape)
df.head()
```

### Identify year columns

`year_cols_orig` contains the value variables (the years). It is dynamically generated by iterating through the DataFrame's columns and selecting those that contain a bracket `[`.

```python
year_cols_orig = [col for col in df.columns if '[' in col]
```

```python

print(year_cols_orig)
df[year_cols_orig].head()
```

### Handle non-standard missing values ('..' to NaN) and convert to float

```python
df[year_cols_orig] = df[year_cols_orig].replace('..', np.nan).astype(float)
```

Replace('..', np.nan): Finds all instances of '..' (which is the World Bank's specific placeholder for missing data) and replaces them with np.nan

### Clean column names to just the year integers for accurate time-series 
E.g., 2000 [YR2000] -> 2000

```python
year_mapping = {col: int(col.split(' ')[0]) for col in year_cols_orig}
df = df.rename(columns=year_mapping)
```

```python
print("Dataset Shape:", df.shape)
print("Time Horizon:", min(year_mapping.values()), "to", max(year_mapping.values()))
```

## 1.3 Handle Missing Values

### Missing Values Analysis

```python
df_ts = df[year_mapping.values()]

missing_per_year = df_ts.isna().sum()
missing_pct_per_year = (missing_per_year / len(df)) * 100
```

Extract only the time series portion (the year columns) for analysis and compute missing value count and percentage per Year.

```python
missing_summary = pd.DataFrame({
    'Missing_Count': missing_per_year,
    'Missing_Percentage (%)': missing_pct_per_year
})

print("\n--- Missing Values Per Year ---")
print(missing_summary)
```

Visualize missing patterns using a Seaborn heatmap.

```python
plt.figure(figsize=(14, 8))

# Dark Navy Blue for "Present" (False/0) and Bright Yellow for "Missing" (True/1)
custom_cmap = sns.color_palette(["#112A46", "#FCE029"])

# The y-axis is the row index (Country/Series), x-axis is the Year
sns.heatmap(df_ts.isna(), cbar=False, cmap=custom_cmap)

# Enhanced typography for the labels
plt.title('Missing Values Heatmap (Yellow = Missing, Dark Blue = Present)', fontsize=16, fontweight='bold', pad=15)
plt.ylabel('Row Index (Unique Country + Series)', fontsize=12)
plt.xlabel('Year Columns', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate x-labels slightly for better readability
plt.tight_layout()
plt.show()
```

Analysis of Missing Patterns:
- Horizontal gaps (Solid yellow lines stretching left to right) indicate specific Country/Series metrics are completely untracked.
- Vertical gaps (like solid yellow bands on specific years) indicate global dataset-wide missing years.
- Trailing missing values on the far right (2022-2024) show data that hasn't been collected or reported yet.

### Preprocessing Data

Convert all data in time-series table to float type

```python
df_ts = df_ts.astype(float)
```

Use a combination of 3 steps:
- Fill in the middle (interpolate).
- Backward fill for the beginning.
- Forward fill for the end.

```python
df_calc = df_ts.interpolate(method='values', axis=1) # Fill in the middle (interpolate)
df_calc = df_calc.bfill(axis=1)                      # Backward fill for the beginning
df_calc = df_calc.ffill(axis=1)                      # Forward fill for the end 
```

Apply to the original table to preserve data: Fill the missing values in the original table using the calculated table above .

```python
id_vars = ["Country Name", "Country Code", "Series Name", "Series Code"]
```

```python
df_time = df_ts.fillna(df_calc).astype(float)
df_final = pd.concat([df[id_vars], df_time], axis=1)
```

Analysis missing value after imputing.

```python
df_ts_final = df_final[year_mapping.values()]

missing_per_year = df_ts_final.isna().sum()
missing_pct_per_year = (missing_per_year / len(df)) * 100
```

```python
missing_summary = pd.DataFrame({
    'Missing_Count': missing_per_year,
    'Missing_Percentage (%)': missing_pct_per_year
})

print("\n--- Missing Values Per Year ---")
print(missing_summary)
```

## Analysis Objective 1: Identifying Growth Leaders and Assessing the Impact of Global Crises (2000 - 2024)

### 1. Indicator Used
* **Data field:** `GDP growth (annual %)`
* **Meaning of the indicator:** This indicator measures the annual percentage growth rate of GDP at constant prices using local currency. It reflects the immediate "health" of an economy, showing whether the economy is expanding or contracting compared to the previous year.

### 2. Reason for Choosing This Indicator
* **Relevance to the objective:** To identify "growth leaders," the annual GDP growth rate is the fairest metric for comparing economies of different sizes. For example, it clearly highlights India’s rapid breakthroughs as well as the more stable performance of larger economies like the United States.
* **Why this indicator:** GDP growth is highly sensitive to external shocks. Tracking this indicator allows us to clearly see the economic "scars" left by major global crises such as the 2008 Global Financial Crisis and the COVID-19 pandemic on each country.

### 3. Chart Type: **Line Chart with Markers and Reference Lines**

### 4. Reason for Choosing This Chart
* **Nature of the data:** The data fluctuates significantly from year to year. A line chart is ideal for visualizing the rhythm of economic cycles, including peaks and troughs over time.

* **Handling real-world data:**
    * **Markers (Points):** These help pinpoint the exact growth rate for each specific year. This is particularly important when comparing crisis years, where growth can drop into negative territory.
    * **Zero-line reference:** A horizontal reference line at $y = 0$ is essential to clearly distinguish between positive growth (even if slow) and actual recession (negative growth).

* **Comparison ability:** The chart makes it easy to observe the contrast between emerging economies and developed economies.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load raw data 
df_raw = pd.read_csv('data.csv') 
df_raw = df_raw.replace(' .. ', np.nan)

gdp_growth_indicator = "GDP growth (annual %)"
df_gdp_base = df_raw[df_raw['Series Name'] == gdp_growth_indicator].copy()
year_cols = [col for col in df_gdp_base.columns if '[' in col]

# Convert data to numeric (float)
df_gdp_base[year_cols] = df_gdp_base[year_cols].apply(pd.to_numeric, errors='coerce')

# Data preparation
# Dataset: Convert to Long format for plotting with Seaborn
df_plot_long = df_gdp_base.melt(id_vars=['Country Name'], value_vars=year_cols, 
                                 var_name='Year_Raw', value_name='Growth_Rate')
df_plot_long['Year'] = df_plot_long['Year_Raw'].str.extract(r'(\d+)')[0].astype(int)

target_countries = [
    'United States', 'China', 'Japan',
    'Germany', 'India', 'United Kingdom', 'France'
]
df_filtered = df_plot_long[df_plot_long['Country Name'].isin(target_countries)]

# plotting
plt.figure(figsize=(15, 8))
sns.set_style("whitegrid")

# Create fixed color palette
palette = sns.color_palette("bright", n_colors=len(target_countries))

# Draw trend lines
sns.lineplot(
    data=df_filtered, 
    x='Year', y='Growth_Rate', hue='Country Name', 
    palette=palette, linewidth=2, alpha=0.8, marker='o', markersize=6
)

# Add reference line y=0
plt.axhline(0, color='black', linestyle='--', linewidth=1.5, alpha=0.7)

# Mark crisis periods
plt.axvspan(2008, 2009, color='gray', alpha=0.2, label='2008 Financial Crisis')
plt.axvspan(2020, 2021, color='red', alpha=0.1, label='COVID-19 Pandemic')

plt.title('GDP Annual Growth Rate: Leaders vs. Global Crises (2000 - 2024)', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP Growth (Annual %)', fontsize=12)
plt.legend(title='Country', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.xticks(range(2000, 2025, 2), fontsize=10)
plt.xlim(2000, 2024)

plt.tight_layout()
plt.show()
```

## Analysis and Conclusions (GDP Growth Trends)

### 1. Key Observations from the Chart

* **Data transparency:** The markers (dots) on the chart show the actual annual GDP growth rates taken from World Bank data.

    * **Complete data series:** Countries like China, India, France, and the United Kingdom have continuous data from 2000 onwards. This helps us clearly see their long-term economic cycles.
    * **Cyclical patterns:** Most countries followed quite similar trends — strong growth between 2003 and 2007, a sharp drop in 2009 due to the global financial crisis, and the lowest point in 2020 because of the COVID-19 pandemic.

* **China and India’s strong performance (Orange and Purple lines):** These two economies stood out as the growth leaders throughout the period.

    * **China:** Grew at a very high rate (over 10%) during the first ten years (2000–2010). After that, the growth slowed down gradually but remained stable at around 5% by 2024.
    * **India:** Showed impressive growth after the pandemic, peaking at nearly 10% in 2021. This really highlights India as one of the most dynamic emerging economies.

* **Impact of the 2008 Financial Crisis (Gray area):** This crisis caused most developed economies to fall into negative growth.

    * **Japan (Green)** and **Germany (Red)** were hit the hardest, with GDP dropping more than -5% in 2009.
    * On the other hand, **China** and **India** still managed to keep positive growth, acting as important stabilizers for the global economy at that time.

* **The COVID-19 Pandemic Shock (Pink area):** This was clearly the biggest shock shown in the entire chart.

    * The **United Kingdom (Brown)** and **France (Pink)** suffered the worst declines, falling to almost -10% in 2020.
    * The **United States (Blue)** also went negative but recovered very quickly with a sharp V-shaped rebound right after.

* **Post-pandemic stabilization (2022–2024):** After the strong recovery in 2021, growth rates across countries have started to converge and become more stable. By 2024, most economies are growing between 0% and 5%.

### 2. Trends and Insights

* **Convergence trend:** The difference in growth rates between countries is getting smaller. Major economies are now entering a phase of more stable but slower growth compared to the early 2000s.

* **Resilience of Emerging Markets:** Emerging economies in Asia, especially China and India, showed much better resilience against global shocks than most Western developed countries.

### 3. Conclusion

From the chart, it is clear that **China and India** have been the main drivers of global economic growth. In contrast, developed countries like the **United States, United Kingdom, and Germany** appear more vulnerable to financial crises and pandemics, often experiencing negative growth during tough times. Overall, the world economy is gradually stabilizing after COVID-19, but growth levels have not yet returned to the high rates seen in the early 2000s.

## Analysis Objective 2: Measuring Economic Convergence and Wealth Inequality through GDP per Capita (2000–2024)

### 1. Indicator Used
* **Data field:** `GDP per capita (current US$)`
* **Meaning of the indicator:** This indicator is calculated by dividing a country’s total GDP by its population in a given year. It shows the average income level per person and is one of the most common ways to measure and compare living standards across countries, even when their population sizes are very different.

### 2. Reason for Choosing This Indicator
* **Relevance to the objective:** While total GDP tells us how big an economy is, GDP per capita helps us understand the “quality” of that growth — how wealthy people actually are on average. This indicator allows us to examine whether developing countries are really catching up with richer nations in terms of income and living standards.
* **Why this indicator:** It clearly highlights the large income gaps between countries (for example, between the United States and India). These gaps reflect differences in labor productivity, technology levels, and overall wealth inequality across the world.

### 3. Chart Type: **Line Chart**

### 4. Reason for Choosing This Chart
* **Nature of the data:** GDP per capita values vary greatly between countries. For instance, the United States is around $80,000 while India is only about $2,500. 

* **Handling real-world data:**
    * **Logarithmic Scale (Log Y-axis):** If we used a normal linear scale, the lines of poorer countries like India or Vietnam would stay very close to the bottom and look almost flat, making it difficult to see their progress. A logarithmic scale helps us better observe the relative growth rates and how fast (or slow) these countries are catching up.
    * **Highlighting Gaps:** The line chart makes it easy to see the persistent gap between high-income, middle-income, and lower-income countries over the 24-year period.

* **Comparison ability:** This chart allows us to clearly identify which countries are successfully narrowing the gap with developed economies and which ones are still struggling to escape the middle-income trap.

```python
# 1. Data preparation
gdp_pc_indicator = "GDP per capita (current US$)"
# Filter the correct indicator from the list
df_pc_base = df_raw[df_raw['Series Name'] == gdp_pc_indicator].copy()

# Identify year columns based on Index list
year_cols = [col for col in df_pc_base.columns if '[' in col]

df_pc_base[year_cols] = df_pc_base[year_cols].apply(pd.to_numeric, errors='coerce')
# Convert to Long format
df_pc_long = df_pc_base.melt(id_vars=['Country Name'], value_vars=year_cols, 
                             var_name='Year_Raw', value_name='GDP_PC')

# Extract year (e.g., '2000 [YR2000]' -> 2000)
df_pc_long['Year'] = df_pc_long['Year_Raw'].str.extract(r'(\d+)')[0].astype(int)

# List of countries exactly matching your data
target_countries = ['United States', 'China', 'Japan', 'Germany', 'India', 'United Kingdom', 'France']
df_pc_filtered = df_pc_long[df_pc_long['Country Name'].isin(target_countries)]

# 2.Plotting
plt.figure(figsize=(15, 8))
sns.set_style("whitegrid")

# Use a color palette that clearly distinguishes countries
palette = sns.color_palette("husl", n_colors=len(target_countries))

# Draw trend lines
sns.lineplot(
    data=df_pc_filtered, 
    x='Year', y='GDP_PC', hue='Country Name', 
    palette=palette, linewidth=2.5, marker='o', markersize=5, alpha=0.8
)

# 3.
for country in target_countries:
    # Filter data for each country and drop NaN to get the actual last value
    country_data = df_pc_filtered[df_pc_filtered['Country Name'] == country].dropna(subset=['GDP_PC'])
    
    if not country_data.empty:
        # Get the row with the largest year that has data
        last_row = country_data.sort_values('Year').iloc[-1]
        last_val = last_row['GDP_PC']
        last_year = last_row['Year']
        
        # Draw value label at the last data point
        plt.text(last_year + 0.3, last_val, f'${last_val:,.0f}', 
                 fontsize=10, fontweight='bold', va='center', color='black')

plt.title('The Growing Wealth Gap: GDP Per Capita Trends (2000 - 2024)', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP per capita (current US$)', fontsize=12)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(range(2000, 2025, 2))
plt.xlim(2000, 2026) # Expand right margin so labels aren't cut off

plt.tight_layout()
plt.show()
```

## Analysis and Conclusions (GDP per Capita Trends)

### 1. Key Observations from the Chart

* **Data transparency:** The markers (dots) on the chart represent the actual GDP per capita values (in current US dollars) for each year.

* **Clear income gaps:** The chart naturally divides the countries into three distinct groups:
    * Very high-income group: the **United States**
    * Developed economies: **Germany, United Kingdom, France, and Japan**
    * Rapidly developing economies: **China and India**

* **Long-term trend:** Almost all countries showed positive growth over the 24-year period, but the speed of improvement varied significantly.

* **United States’ absolute dominance (Pink line):** The United States maintained its leading position with no real competitor.
    * Starting from about $36,000 in 2000, the US experienced strong growth and reached a record high of $84,534 in 2024.
    * The gap between the US and the second-place country (Germany) has continued to widen, highlighting America’s superior ability to sustain income growth.

* **European countries and Japan’s stagnation:** The traditional developed economies showed noticeable differences in performance.
    * **Germany (Green), United Kingdom (Blue), and France (Light Pink)** maintained relatively stable growth and stayed close to each other, ranging between $46,000 and $56,000. Germany currently leads this group.
    * **Japan (Olive Green)** was a special case. After 2012, its GDP per capita began to decline slightly and then stagnated, ending at $32,487 in 2024 — significantly lower than other G7 countries.

* **China and India’s growth (Orange and Teal lines):** Although their absolute values are still much lower, their relative growth has been impressive.
    * **China:** Made remarkable progress, rising from nearly zero in 2000 to $13,303 in 2024. This represents a strong transition from low-income to upper-middle-income status.
    * **India:** Has entered a phase of more stable growth, reaching $2,695 in 2024. While it remains the lowest in the group, its long-term growth potential is clearly visible.

### 2. Trends and Insights

* **The Growing Wealth Gap:** Contrary to the idea of convergence, the gap between the richest country (the United States) and the others has actually widened over time, especially compared to the G7 countries and Japan.

* **Asian Emergence:** China’s rapid improvement in living standards stands out as the brightest point in the chart. Meanwhile, Japan has been facing prolonged stagnation in per capita income.

### 3. Conclusion

The chart clearly confirms the United States’ position as an economic superpower, with a GDP per capita far higher than any other country in the group. While **China** has been closing the gap with developed nations at an impressive pace, **Japan** appears to be falling behind its Western counterparts in terms of income levels. For **India**, the biggest challenge in the coming decades will be to maintain strong growth and successfully break out of the low-income group.

## Analysis Objective 3: Comparing Deindustrialization Trends Across Major Economies (2000–2024)

### 1. Indicator Used
* **Data field:** `Industry (including construction), value added (% of GDP)`
* **Meaning of the indicator:** This indicator shows the share of the industrial sector (including manufacturing, construction, mining, and utilities) in a country’s total GDP. It helps us understand how important industry is to the overall economy and the level of industrialization of each country.

### 2. Reason for Choosing This Indicator
* **Relevance to the objective:** To see which countries are experiencing deindustrialization and how fast it is happening, we need to track the changing share of industry in GDP over time.
* **Why this indicator:** A decreasing percentage usually indicates deindustrialization. This process is often linked to:
    * The shift toward service-based economies
    * Moving manufacturing activities to other countries (offshoring)
    * Broader structural changes in the economy

### 3. Chart Type: **Line Chart with Markers**

### 4. Reason for Choosing This Chart
* **Nature of the data:** Since this is time-series data spanning 24 years, a line chart is the most suitable way to observe long-term structural changes in the economy.

* **Handling real-world data:**
    * **Markers (dots):** These clearly show the actual data points each year, making it easy to distinguish real values from the trend lines.
    * **Trend lines:** Help us see whether the decline in industrial share is happening gradually or sharply.

* **Comparison ability:** With multiple countries on the same chart, we can easily compare:
    * The current level of industrialization (who still relies more on industry)
    * The speed of deindustrialization (whose industrial share is falling the fastest)

```python
industry_indicator = "Industry (including construction), value added (% of GDP)"
df_ind_base = df_raw[df_raw['Series Name'] == industry_indicator].copy()

year_cols = [col for col in df_ind_base.columns if '[' in col]

# Convert safely to numeric
df_ind_base[year_cols] = df_ind_base[year_cols].apply(pd.to_numeric, errors='coerce')

# Convert to long format
df_ind_long = df_ind_base.melt(
    id_vars=['Country Name'], 
    value_vars=year_cols,
    var_name='Year_Raw', 
    value_name='Industry_Percent'
)

df_ind_long['Year'] = df_ind_long['Year_Raw'].str.extract(r'(\d+)')[0].astype(int)

# Filter countries
target_countries = [
    'United States', 'China', 'Japan',
    'Germany', 'India', 'United Kingdom', 'France'
]

df_ind_filtered = df_ind_long[df_ind_long['Country Name'].isin(target_countries)]

# 2.
plt.figure(figsize=(15, 8))
sns.set_style("whitegrid")

palette = sns.color_palette("tab10", n_colors=len(target_countries))

sns.lineplot(
    data=df_ind_filtered, 
    x='Year', y='Industry_Percent', hue='Country Name',
    palette=palette, linewidth=2.5, marker='o', markersize=5
)

# Titles
plt.title('Deindustrialization Trends in Major Economies (2000–2024)', 
          fontsize=16, fontweight='bold')

plt.xlabel('Year')
plt.ylabel('Industry Value Added (% of GDP)')
plt.legend(title='Country', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.xticks(range(2000, 2025, 2))
plt.xlim(2000, 2024)

plt.tight_layout()
plt.show()
```

## Analysis and Conclusions

### 1. Key Observations from the Chart

* **Data transparency:** The lines show the share of industry (including construction) as a percentage of total GDP for each country. The markers (dots) indicate the actual annual values from the data.

* **China’s structural shift (Orange line):** China recorded the most dramatic change. Although it still has the highest share (above 35% in 2024), its industrial contribution has declined from a peak of nearly 47% in 2006.
* **The Manufacturing Leaders (Germany vs. China vs. USA):**
    * **Germany (Red line):** Germany has shown remarkable stability, maintaining its industrial share around 25–27%. Compared to other Western countries, Germany has successfully preserved a strong industrial base, reinforcing its position as the “factory of Europe.”
    * **China (Orange line):** Even though its share is decreasing, it remains roughly double that of the United States, highlighting the central role of manufacturing in China’s economy.
    * **United States (Blue line):** The US has a relatively low industrial share that continues to decline slightly, now below 18%. America completed its shift toward a service-based economy a long time ago.

* **India’s volatility (Purple line):** India’s industrial share has been quite unstable. After rising to around 31% during 2006–2010, it gradually declined to about 25% by 2024. This shows the challenges India has faced in pushing forward its “Make in India” initiative.

* **European Group (UK & France):** Both the **United Kingdom (Brown)** and **France (Pink)** have the lowest industrial shares in the group (below 18%), reflecting a deep process of deindustrialization in these countries.

### 2. Trends and Insights

* **Fastest decline:** China experienced the fastest drop in industrial share over the past 15 years (more than 10 percentage points from its peak).

* **Polarization in manufacturing:** There is a clear divide between the “manufacturing group” (Germany, China, and Japan — all staying above 25%) and the “service-oriented group” (United States, United Kingdom, and France — below 20%). Germany’s stability demonstrates that maintaining high-quality industry can effectively resist full deindustrialization.

### 3. Conclusion

The chart confirms that a significant structural shift away from industry is taking place, most notably in China. While countries like the **United States, United Kingdom, and France** have long settled into low industrial shares, **Germany** remains a rare exception by successfully preserving a strong and stable industrial foundation. For developing countries such as **India**, maintaining a stable industrial share continues to be a major challenge in the face of intense global competition.

## Analysis Objective 4: Evaluating the Structural Transformation of Agriculture in India Compared to Major Economies (2000–2024)

### 1. Indicator Used
* **Data field:** `Agriculture, forestry, and fishing, value added (% of GDP)`
* **Meaning of the indicator:** This indicator measures the contribution of the agricultural sector (including farming, forestry, and fishing) to a country’s total GDP. It shows how much an economy still depends on primary sector activities and serves as an important signal of its stage of economic development.

### 2. Reason for Choosing This Indicator
* **Relevance to the objective:** This analysis focuses on how the role of agriculture in India’s economy has changed over time compared to other major economies. ## Analysis Objective 5: Service Sector Dominance and Structural Maturity

### 1. Indicator Used
* **Data field:** `Services, value added (% of GDP)`
* **Meaning of the indicator:** This measures the net output of the service sector (including government, financial services, healthcare, etc.) as a percentage of a country's total GDP.

### 2. Reason for Choosing This Indicator
* **Relevance to the objective:** It is the primary metric used to determine the "maturity" of an economy; a higher percentage usually indicates a shift toward a post-industrial society.
* **Why this indicator:** It allows us to distinguish between industrial powerhouses and service-based economies within the group over the 2000–2024 period.

### 3. Chart Type: **Boxenplot & Lineplot**

### 4. Reason for Choosing This Chart
* **Nature of the data:** A **Boxenplot** is superior for showing the distribution, median, and outliers across different countries.
* **Clarity:** A supplementary **Lineplot** is used to visualize the specific trajectory of growth for each nation, making it easy to spot the "strongest increase."

```python
#Load data
df_imputed = pd.read_csv("data_imputed.csv")

# Filter indicator
service_indicator = "Services, value added (% of GDP)"
df_services = df_imputed[df_imputed['Series Name'] == service_indicator].copy()

# Melt to long format for Seaborn
year_cols = [str(y) for y in range(2000, 2025)]
df_services_long = df_services.melt(id_vars=['Country Name'], value_vars=year_cols, 
                                    var_name='Year', value_name='Services_pct')

import matplotlib.pyplot as plt
import seaborn as sns

# 1. Tính toán thứ tự các nước theo trung bình % Dịch vụ để tạo sự liên kết
order = df_services_long.groupby("Country Name")["Services_pct"].median().sort_values(ascending=False).index

# 2. Thiết lập bảng màu đồng nhất cho tất cả các nước dựa trên thứ tự này
palette = sns.color_palette("viridis", len(order))
color_map = dict(zip(order, palette))

# 3. Tạo Figure
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 12))

# --- BIỂU ĐỒ 1: BOXENPLOT (Dùng order và color_map) ---
sns.boxenplot(data=df_services_long, x='Country Name', y='Services_pct', 
              order=order, palette=color_map, ax=ax1)
ax1.set_title('1. Structural Maturity: Distribution of Services % GDP (Ordered by Median)', fontsize=14, fontweight='bold')
ax1.set_ylabel('Services % of GDP')
ax1.tick_params(axis='x', rotation=30)

# --- BIỂU ĐỒ 2: LINEPLOT (Dùng color_map để cùng màu với biểu đồ trên) ---
sns.lineplot(data=df_services_long, x='Year', y='Services_pct', 
             hue='Country Name', hue_order=order, palette=color_map, 
             marker='o', linewidth=2, ax=ax2)
ax2.set_title('2. Growth Trajectory: Evolution of Service Sectors (2000-2024)', fontsize=14, fontweight='bold')
ax2.set_ylabel('Services % of GDP')
ax2.legend(title='Country (High to Low)', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

## Analysis Objective 9: The Correlation Between Industrial Strength and GDP Growth

### 1. Indicator Used
* **Data fields:** `GDP growth (annual %)` and `Industry, value added (% of GDP)`
* **Meaning of the indicators:** GDP growth measures the speed of economic expansion, while Industry % GDP measures the weight of manufacturing and construction.

### 2. Reason for Choosing This Indicator
* **Relevance to the objective:** This determines if countries with larger industrial bases experience faster (or more volatile) growth rates.
* **Why this indicator:** To investigate if the "industrial engine" model still holds true for these nations in the modern era.

### 3. Chart Type: **Regression Plot (Regplot)**

### 4. Reason for Choosing This Chart
* **Nature of the data:** A **Regplot** combines a scatter plot with a trend line, making the relationship between two continuous variables visible.
* **Clarity:** Including the Pearson correlation coefficient (r) provides a mathematical basis for the conclusion.

```python
!pip install spicy
```

```python
from scipy import stats

# Extract and align two indicators
df_growth = df_imputed[df_imputed['Series Name'] == "GDP growth (annual %)"].set_index('Country Name')[year_cols].stack().reset_index()
df_ind = df_imputed[df_imputed['Series Name'] == "Industry (including construction), value added (% of GDP)"].set_index('Country Name')[year_cols].stack().reset_index()

# Combine
df_corr = pd.DataFrame({
    'Growth': df_growth[0].values,
    'Industry': df_ind[0].values
})

# Calculate Pearson Correlation
r, p_value = stats.pearsonr(df_corr['Industry'], df_corr['Growth'])

# Plot
plt.figure(figsize=(10, 6))
sns.regplot(data=df_corr, x='Industry', y='Growth', 
            scatter_kws={'alpha':0.4, 'color':'#3498db'}, 
            line_kws={'color':'red', 'label': f'Linear Fit (r={r:.2f})'})

# Add Correlation Coefficient text onto the plot
plt.text(16, 13, f'Pearson r = {r:.2f}\np-value < 0.001', 
         fontsize=12, fontweight='bold', color='red', 
         bbox=dict(facecolor='white', alpha=0.5))

plt.title('Correlation: Industrial Contribution vs. Annual GDP Growth', fontsize=14, fontweight='bold')
plt.xlabel('Industry (% of GDP)')
plt.ylabel('Annual GDP Growth (%)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### **ANALYSIS AND CONCLUSIONS (OBJECTIVE 9)**
**Objective 9: Correlation between Industrial Contribution and Annual GDP Growth**

**1. Key observations from the chart**
* **Significant Positive Correlation:**
    * The Pearson correlation coefficient (**r = 0.59**) indicates a **moderate-to-strong positive relationship** between industrial value-added and annual GDP growth.
    * The p-value (**< 0.001**) confirms that this relationship is statistically highly significant, meaning the observed trend is very unlikely to be due to random chance.
* **The "Industrial Sweet Spot" for High Growth:**
    * Data points in the high-industry zone (**Industry > 40%**) consistently coincide with exceptional GDP growth rates exceeding **8% to 12%**. This cluster likely represents periods of rapid industrialization.
* **Stability vs. Volatility:**
    * Economies with an industrial share between **15% and 25%** (typically G7 nations) show a dense cluster of stable growth between **0% and 4%**.
    * However, as the industrial share moves into the **25% - 32%** range, volatility increases significantly, with growth rates ranging from **-6% to nearly 10%**.
* **Post-Crisis Recovery Points:**
    * The few negative growth outliers (e.g., around **-10%**) suggest that even with a decent industrial base, external shocks can still impact performance, though the overall trend remains firmly positive.

**2. Trends and insights**
* **Industrial Sector as a Growth Engine:** The upward slope of the red regression line proves that for these 7 major economies, a larger industrial sector acts as a powerful driver of economic expansion.

## Analysis Objective 10: Trade Openness as a Pathway to National Wealth

### 1. Indicator Used
* **Data fields:** `Exports of goods and services (% of GDP)` and `GDP per capita (current US$)`
* **Meaning of the indicators:** One represents trade intensity; the other represents the average economic output per person (a proxy for wealth).

### 2. Reason for Choosing This Indicator
* **Relevance to the objective:** It answers whether "selling more to the world" actually makes the average citizen richer in these specific countries.
* **Why this indicator:** It links a macro-economic strategy (exports) to a micro-economic outcome (per capita wealth) across different developmental stages.

### 3. Chart Type: **Scatter Plot (with Log Scale)**

### 4. Reason for Choosing This Chart
* **Nature of the data:** **Scatter plots** are ideal for identifying clusters and trends among different countries.
* **Clarity:** Using a **Log Scale** for GDP per capita is necessary to manage the wide disparity in wealth between the nations.

```python
# Extract 2024 data
df_exp_24 = df_imputed[df_imputed['Series Name'] == "Exports of goods and services (% of GDP)"][['Country Name', '2024']]
df_pc_24 = df_imputed[df_imputed['Series Name'] == "GDP per capita (current US$)"][['Country Name', '2024']]

# Merge
df_wealth = pd.merge(df_exp_24, df_pc_24, on='Country Name')
df_wealth.columns = ['Country', 'Exports', 'GDP_per_Capita']

# Plot
plt.figure(figsize=(12, 7))
sns.scatterplot(data=df_wealth, x='Exports', y='GDP_per_Capita', hue='Country', s=200, palette='deep')
plt.yscale('log')
plt.title('Trade Openness (2024) vs. National Wealth (Log Scale)', fontsize=14, fontweight='bold')
plt.grid(True, which="both", ls="--", alpha=0.5)
plt.show()
```

**Objective 10: Trade Openness (2024) vs. National Wealth (Log Scale)**

**1. Key observations from the chart**
* **The High-Income Cluster (G7):**
    * There is a distinct cluster at the top of the chart representing the wealthiest nations (**USA, Germany, UK, France, Japan**), all maintaining a GDP per capita above **$30,000** (log scale).
    * **The United States** is the primary outlier in this group: it has the highest wealth (near **$10^5$**) but the lowest trade openness (Exports ~11%), suggesting its wealth is driven by a massive internal economy.
* **The "Export-Led" Wealth Model (Germany):**
    * **Germany** sits at the far right of the chart, showing the highest trade openness (over **40%**) while maintaining a very high GDP per capita. This confirms that Germany successfully converted its export strategy into national prosperity.
* **The Emerging Market Gap:**
    * **India and China** occupy the lower section of the chart. While their trade openness (Exports ~20%) is similar to Japan's, their GDP per capita remains significantly lower (between **$10^3$ and $1.5 \times$ $10^4$**).
    * **India** is currently at the bottom of this group, highlighting the vast potential for per-capita wealth growth as it integrates further into global markets.

**2. Trends and insights**

* **Internal Market vs. Trade Reliance:** The chart illustrates two different paths to wealth: the **US model** (Internal Market focus) vs. the **German/UK model** (Global Trade focus). Both lead to high national wealth, but through different structural dependencies.

The analysis confirms that while high trade openness (Exports % GDP) often correlates with national wealth in Europe, it is not the only path to prosperity, as evidenced by the United States. For emerging giants like **India and China**, the challenge lies in translating their trade volumes into higher per-capita income to bridge the gap with the G7 nations.
