### Load the dataset

```python
df = pd.read_csv("data.csv")
print(df.shape)
df.head()
```

### Identify year columns

`year_cols_orig` contains the value variables (the years). It is dynamically generated by iterating through the DataFrame's columns and selecting those that contain a bracket `[`.

```python
year_cols_orig = [col for col in df.columns if '[' in col]
```

```python

print(year_cols_orig)
df[year_cols_orig].head()
```

### Handle non-standard missing values ('..' to NaN) and convert to float

```python
df[year_cols_orig] = df[year_cols_orig].replace('..', np.nan).astype(float)
```

Replace('..', np.nan): Finds all instances of '..' (which is the World Bank's specific placeholder for missing data) and replaces them with np.nan

### Clean column names to just the year integers for accurate time-series 
E.g., 2000 [YR2000] -> 2000

```python
year_mapping = {col: int(col.split(' ')[0]) for col in year_cols_orig}
df = df.rename(columns=year_mapping)
```

```python
print("Dataset Shape:", df.shape)
print("Time Horizon:", min(year_mapping.values()), "to", max(year_mapping.values()))
```

```python
df.head()
```

## 1.3 Handle Missing Values

### Missing Values Analysis

```python
df_ts = df[year_mapping.values()]

missing_per_year = df_ts.isna().sum()
missing_pct_per_year = (missing_per_year / len(df)) * 100
```

Extract only the time series portion (the year columns) for analysis and compute missing value count and percentage per Year.

```python
missing_summary = pd.DataFrame({
    'Missing_Count': missing_per_year,
    'Missing_Percentage (%)': missing_pct_per_year
})

print("\n--- Missing Values Per Year ---")
print(missing_summary)
```

Visualize missing patterns using a Seaborn heatmap.

```python
plt.figure(figsize=(14, 8))

# Dark Navy Blue for "Present" (False/0) and Bright Yellow for "Missing" (True/1)
custom_cmap = sns.color_palette(["#112A46", "#FCE029"])

# The y-axis is the row index (Country/Series), x-axis is the Year
sns.heatmap(df_ts.isna(), cbar=False, cmap=custom_cmap)

# Enhanced typography for the labels
plt.title('Missing Values Heatmap (Yellow = Missing, Dark Blue = Present)', fontsize=16, fontweight='bold', pad=15)
plt.ylabel('Row Index (Unique Country + Series)', fontsize=12)
plt.xlabel('Year Columns', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate x-labels slightly for better readability
plt.tight_layout()
plt.show()
```

Analysis of Missing Patterns:
- Horizontal gaps (Solid yellow lines stretching left to right) indicate specific Country/Series metrics are completely untracked.
- Vertical gaps (like solid yellow bands on specific years) indicate global dataset-wide missing years.
- Trailing missing values on the far right (2022-2024) show data that hasn't been collected or reported yet.

### Preprocessing Data

Convert all data in time-series table to float type

```python
df_ts = df_ts.astype(float)
```

Use a combination of 3 steps:
- Fill in the middle (interpolate).
- Backward fill for the beginning.
- Forward fill for the end.

```python
df_calc = df_ts.interpolate(method='values', axis=1) # Fill in the middle (interpolate)
df_calc = df_calc.bfill(axis=1)                      # Backward fill for the beginning
df_calc = df_calc.ffill(axis=1)                      # Forward fill for the end 
```

Apply to the original table to preserve data: Fill the missing values in the original table using the calculated table above .

```python
id_vars = ["Country Name", "Country Code", "Series Name", "Series Code"]
```

```python
df_time = df_ts.fillna(df_calc).astype(float)
df_final = pd.concat([df[id_vars], df_time], axis=1)
```

Analysis missing value after imputing.

```python
df_ts_final = df_final[year_mapping.values()]

missing_per_year = df_ts_final.isna().sum()
missing_pct_per_year = (missing_per_year / len(df)) * 100
```

```python
missing_summary = pd.DataFrame({
    'Missing_Count': missing_per_year,
    'Missing_Percentage (%)': missing_pct_per_year
})

print("\n--- Missing Values Per Year ---")
print(missing_summary)
```

## Analysis Objective 5: Service Sector Dominance and Structural Maturity

### 1. Indicator Used
* **Data field:** `Services, value added (% of GDP)`
* **Meaning of the indicator:** This measures the net output of the service sector (including government, financial services, healthcare, etc.) as a percentage of a country's total GDP.

### 2. Reason for Choosing This Indicator
* **Relevance to the objective:** It is the primary metric used to determine the "maturity" of an economy; a higher percentage usually indicates a shift toward a post-industrial society.
* **Why this indicator:** It allows us to distinguish between industrial powerhouses and service-based economies within the group over the 2000–2024 period.

### 3. Chart Type: **Boxenplot & Lineplot**

### 4. Reason for Choosing This Chart
* **Nature of the data:** A **Boxenplot** is superior for showing the distribution, median, and outliers across different countries.
* **Clarity:** A supplementary **Lineplot** is used to visualize the specific trajectory of growth for each nation, making it easy to spot the "strongest increase."

```python
#Load data
df_imputed = pd.read_csv("data_imputed.csv")

# Filter indicator
service_indicator = "Services, value added (% of GDP)"
df_services = df_imputed[df_imputed['Series Name'] == service_indicator].copy()

# Melt to long format for Seaborn
year_cols = [str(y) for y in range(2000, 2025)]
df_services_long = df_services.melt(id_vars=['Country Name'], value_vars=year_cols, 
                                    var_name='Year', value_name='Services_pct')

import matplotlib.pyplot as plt
import seaborn as sns

# 1. Tính toán thứ tự các nước theo trung bình % Dịch vụ để tạo sự liên kết
order = df_services_long.groupby("Country Name")["Services_pct"].median().sort_values(ascending=False).index

# 2. Thiết lập bảng màu đồng nhất cho tất cả các nước dựa trên thứ tự này
palette = sns.color_palette("viridis", len(order))
color_map = dict(zip(order, palette))

# 3. Tạo Figure
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 12))

# --- BIỂU ĐỒ 1: BOXENPLOT (Dùng order và color_map) ---
sns.boxenplot(data=df_services_long, x='Country Name', y='Services_pct', 
              order=order, palette=color_map, ax=ax1)
ax1.set_title('1. Structural Maturity: Distribution of Services % GDP (Ordered by Median)', fontsize=14, fontweight='bold')
ax1.set_ylabel('Services % of GDP')
ax1.tick_params(axis='x', rotation=30)

# --- BIỂU ĐỒ 2: LINEPLOT (Dùng color_map để cùng màu với biểu đồ trên) ---
sns.lineplot(data=df_services_long, x='Year', y='Services_pct', 
             hue='Country Name', hue_order=order, palette=color_map, 
             marker='o', linewidth=2, ax=ax2)
ax2.set_title('2. Growth Trajectory: Evolution of Service Sectors (2000-2024)', fontsize=14, fontweight='bold')
ax2.set_ylabel('Services % of GDP')
ax2.legend(title='Country (High to Low)', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

### **ANALYSIS AND CONCLUSIONS (OBJECTIVE 5)**
**Objective 5: Service Sector Dominance and Structural Maturity (2000–2024)**

**1. Key observations from the charts**
* **The "Service Giants" Hierarchy (Chart 1):**
    * **The United States** maintains the highest structural maturity, with its service sector consistently hovering between **73% and 78%** of GDP. 
    * **The UK, France, and Japan** form a high-stability cluster, all maintaining a service base above **65%**. Their narrow boxenplots indicate very low volatility in economic structure over the 24-year period.
* **Rapid Structural Transformation (Chart 2):**
    * **China** shows the most aggressive "upward climb" in the line plot. It started at the lowest point in 2000 (**~40%**) and has surged to nearly **57%** by 2024, overtaking India in terms of structural transition speed.
    * **India** also shows a steady rise, though its trajectory is more gradual compared to China's recent surge, fluctuating around the **45-50%** mark.
* **The "Industrial Anchor" (Germany):**
    * **Germany** is an interesting case among developed nations; it has a lower service percentage (**~62-64%**) compared to its G7 peers. This confirms Germany's unique economic model, which deliberately retains a much stronger industrial base.

**2. Trends and insights**
* **Post-Industrial Convergence:** The line plot reveals a narrowing gap between the emerging giants (China/India) and the developed G7 nations. While the G7 has reached a "saturation point" in services, the emerging nations are still in the high-growth phase of structural transition.
* **Economic Maturity vs. Stability:** The Boxenplot shows that as the service sector grows, the "boxes" become narrower (e.g., USA, UK). This suggests that high-service economies experience more stable structural distributions compared to the wider, more volatile distributions seen in China.

**3. Conclusion**
The analysis confirms a global shift toward services as the primary engine of GDP. While the **USA** remains the benchmark for a post-industrial service economy, **China** is undergoing the fastest structural modernization in the group, signaling a pivot away from its traditional identity as a purely manufacturing-based economy.

## Analysis Objective 6: Export Performance in the Era of Trade Tensions (2018–2024)

### 1. Indicator Used
* **Data field:** `Exports of goods and services (% of GDP)`
* **Meaning of the indicator:** It represents the value of all goods and market services provided to the rest of the world as a share of GDP.

### 2. Reason for Choosing This Indicator
* **Relevance to the objective:** It measures a country's "Trade Openness" and its reliance on foreign markets for economic survival.
* **Why this indicator:** Focusing on the post-2018 period reveals which nations remained resilient or capitalized on shifting trade routes during the trade war era.

### 3. Chart Type: **Barplot**

### 4. Reason for Choosing This Chart
* **Nature of the data:** A **Barplot** with an estimator (mean) clearly ranks countries by their average performance in a specific recent timeframe.
* **Clarity:** It provides a direct comparison of export intensity among the nations during the trade war period.

```python
# Filter indicator
export_indicator = "Exports of goods and services (% of GDP)"
df_exports = df_imputed[df_imputed['Series Name'] == export_indicator].copy()

# Average for 2018-2024 (Trade war era)
post_2018_cols = [str(y) for y in range(2018, 2025)]
df_exports['Avg_Post_2018'] = df_exports[post_2018_cols].mean(axis=1)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=df_exports.sort_values('Avg_Post_2018', ascending=False), 
            x='Country Name', y='Avg_Post_2018', palette='rocket')
plt.title('Average Exports % GDP (Trade War Era: 2018 - 2024)', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.ylabel('Mean Exports % GDP')
plt.show()
```

### **ANALYSIS AND CONCLUSIONS (OBJECTIVE 6)**
**Objective 6: Export Performance in the Era of Trade Tensions (2018–2024)**

**1. Key observations from the chart**
* **Germany’s Structural Superiority:**
    * **Germany** consistently maintains the highest export-to-GDP ratio, fluctuating between **43% and 51%**. 
    * Despite a sharp "V-shaped" dip around 2020, it recovered rapidly, far outperforming other major economies in trade intensity.
* **The European Convergence (UK & France):**
    * **The UK and France** follow a very similar trajectory, with both nations sitting in the **26% to 35%** range. 
    * Interestingly, the **UK** saw a significant surge post-2021, eventually overtaking France in export contribution by 2024.
* **The "Internal Giants" (China, India, USA):**
    * **China and India** exhibit nearly identical trends, hovering between **17% and 21%**. Both showed resilience with a slight upward tick after 2020.
    * **The United States** remains at the bottom of the group with a very flat line around **10-12%**, showing the least sensitivity to global trade fluctuations.
* **Japan’s Moderate Growth:**
    * **Japan** shows a steady, gradual increase from **~14% in 2020 to nearly 22% in 2024**, moving past the baseline of India and China.

**2. Trends and insights**
* **The 2020 Global Trade Shock:** All nations (except perhaps the USA) experienced a visible dip in 2020 due to pandemic lockdowns, but the recovery across the G7 and emerging powers was remarkably synchronized.
* **Post-Pandemic Expansion:** From 2021 to 2024, there is a general upward trend for almost all countries, suggesting that global trade integration intensified rather than weakened during the recent era of tensions.
* **Strategic Autonomy:** The low and stable ratio of the **USA** compared to the high volatility of **Germany** highlights the difference between a consumption-led domestic economy and an export-led industrial economy.

**3. Conclusion**
The data proves that **Germany** remains the world's most trade-dependent major power. While the **UK** is showing newfound export momentum, the **USA, China, and India** continue to operate as "Internal Engines," where domestic activity far outweighs the importance of foreign trade.

## Analysis Objective 8: Structural Transformation of the Economy (2000 vs 2024)

### 1. Indicator Used
* **Data fields:** `Agriculture % GDP`, `Industry % GDP`, `Services % GDP`
* **Meaning of the indicators:** These three sectors represent the total value added to the economy, traditionally summing up to nearly 100%.

### 2. Reason for Choosing This Indicator
* **Relevance to the objective:** To see if countries are "industrializing" (moving from Agri to Industry) or "modernizing" (moving to Services).
* **Why this indicator:** Comparing the 2000-2010 period with 2015-2024 highlights the long-term structural shifts in each nation's "economic DNA."

### 3. Chart Type: **Grouped Bar Chart (FacetGrid)**

### 4. Reason for Choosing This Chart
* **Nature of the data:** Using a **FacetGrid** of bar charts allows us to compare three different variables (sectors) across multiple time points.
* **Clarity:** It visualizes the internal "re-balancing" of each economy side-by-side.

```python
# Filter the three sectors
sectors = ["Agriculture, forestry, and fishing, value added (% of GDP)",
           "Industry (including construction), value added (% of GDP)",
           "Services, value added (% of GDP)"]

df_struct = df_imputed[df_imputed['Series Name'].isin(sectors)].copy()

# Melt and Pivot
df_struct_melt = df_struct.melt(id_vars=['Country Name', 'Series Name'], value_vars=['2000', '2024'], var_name='Year')
df_plot_struct = df_struct_melt.pivot_table(index=['Country Name', 'Year'], columns='Series Name', values='value').reset_index()

# Rename for easier plotting
df_plot_struct.columns = ['Country', 'Year', 'Agri', 'Ind', 'Serv']
df_plot_struct_melt = df_plot_struct.melt(id_vars=['Country', 'Year'], value_vars=['Agri', 'Ind', 'Serv'])

# Plot
g = sns.catplot(data=df_plot_struct_melt, x='Year', y='value', hue='variable', col='Country', 
                kind='bar', col_wrap=3, height=4, palette='Set2')

# --- ĐOẠN SỬA ĐỂ HIỆN NHÃN ĐỒNG ĐỀU ---
# Duy trì nhãn cho tất cả các biểu đồ con
for ax in g.axes.flat:
    ax.tick_params(labelbottom=True) # Hiện 2000, 2024
    ax.set_xlabel('Year')            # Hiện chữ Year

g.set_axis_labels("", "Percentage of GDP") # Xóa nhãn chung ở ngoài để tránh trùng lặp
g.set_titles("{col_name}")
plt.subplots_adjust(top=0.85, hspace=0.3) # Tăng khoảng cách hspace để không bị đè chữ
g.fig.suptitle('Structural Shift: 2000 vs 2024', fontsize=16, fontweight='bold')
plt.show()
```

### **ANALYSIS AND CONCLUSIONS (OBJECTIVE 8)**
**Objective 8: Structural Transformation of the Economy (2000 vs 2024)**

**1. Key observations from the chart**
* **Universal Dominance of the Service Sector (Serv):**
    * By **2024**, the **Services sector (blue/grey bars)** has solidified its position as the primary driver of GDP in all seven nations.
    * The **United States, United Kingdom, and France** exhibit the highest service-dependence, with their service sectors contributing over **70%** of total GDP.
* **Significant Shrinkage of Agriculture (Agri):**
    * **India and China** show the most dramatic decline in **Agriculture (green bars)**. In India, Agriculture dropped from over **20% in 2000** to approximately **15% in 2024**, while China saw its agricultural share nearly halved.
* **Industrial Rebalancing (Ind):**
    * **China** remains the industrial heavyweight of the group, though its **Industry (orange bars)** share decreased from around **45% in 2000** to roughly **38% in 2024** as it pivoted toward services.
    * **Germany and Japan** continue to maintain a robust industrial base (above **25%**), highlighting their persistent roles as global manufacturing hubs.

**2. Trends and insights**
* **Structural Convergence:** Comparing the **2000 and 2024** data across the "Year" axis reveals a clear "convergence" trend. Eastern economies (China and India) are rapidly evolving to match the "Service-heavy" economic DNA of Western G7 powers.
* **Modernization through De-agrarianization:** The movement from 2000 to 2024 illustrates a global transition where national wealth is increasingly decoupled from primary production (Agriculture) and reallocated toward intangible value and high-tech services.
* **Stability of the G7 Model:** While China and India are undergoing radical metamorphosis, the **USA, UK, and France** maintain a highly stable, service-dominant structure with only minor incremental shifts over the 24-year period.

**3. Conclusion**
The comparison between 2000 and 2024 confirms that a transition to a service-led economy is the global standard for economic maturity. **China** stands out as the nation undergoing the most intense structural shift, successfully moving from an industrial-heavy model toward a more balanced, service-oriented economy.

```python
from scipy import stats

# Extract and align two indicators
df_growth = df_imputed[df_imputed['Series Name'] == "GDP growth (annual %)"].set_index('Country Name')[year_cols].stack().reset_index()
df_ind = df_imputed[df_imputed['Series Name'] == "Industry (including construction), value added (% of GDP)"].set_index('Country Name')[year_cols].stack().reset_index()

# Combine
df_corr = pd.DataFrame({
    'Growth': df_growth[0].values,
    'Industry': df_ind[0].values
})

# Calculate Pearson Correlation
r, p_value = stats.pearsonr(df_corr['Industry'], df_corr['Growth'])

# Plot
plt.figure(figsize=(10, 6))
sns.regplot(data=df_corr, x='Industry', y='Growth', 
            scatter_kws={'alpha':0.4, 'color':'#3498db'}, 
            line_kws={'color':'red', 'label': f'Linear Fit (r={r:.2f})'})

# Add Correlation Coefficient text onto the plot
plt.text(16, 13, f'Pearson r = {r:.2f}\np-value < 0.001', 
         fontsize=12, fontweight='bold', color='red', 
         bbox=dict(facecolor='white', alpha=0.5))

plt.title('Correlation: Industrial Contribution vs. Annual GDP Growth', fontsize=14, fontweight='bold')
plt.xlabel('Industry (% of GDP)')
plt.ylabel('Annual GDP Growth (%)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

## Analysis Objective 10: Trade Openness as a Pathway to National Wealth

### 1. Indicator Used
* **Data fields:** `Exports of goods and services (% of GDP)` and `GDP per capita (current US$)`
* **Meaning of the indicators:** One represents trade intensity; the other represents the average economic output per person (a proxy for wealth).

### 2. Reason for Choosing This Indicator
* **Relevance to the objective:** It answers whether "selling more to the world" actually makes the average citizen richer in these specific countries.
* **Why this indicator:** It links a macro-economic strategy (exports) to a micro-economic outcome (per capita wealth) across different developmental stages.

### 3. Chart Type: **Scatter Plot (with Log Scale)**

### 4. Reason for Choosing This Chart
* **Nature of the data:** **Scatter plots** are ideal for identifying clusters and trends among different countries.
* **Clarity:** Using a **Log Scale** for GDP per capita is necessary to manage the wide disparity in wealth between the nations.

```python
# Extract 2024 data
df_exp_24 = df_imputed[df_imputed['Series Name'] == "Exports of goods and services (% of GDP)"][['Country Name', '2024']]
df_pc_24 = df_imputed[df_imputed['Series Name'] == "GDP per capita (current US$)"][['Country Name', '2024']]

# Merge
df_wealth = pd.merge(df_exp_24, df_pc_24, on='Country Name')
df_wealth.columns = ['Country', 'Exports', 'GDP_per_Capita']

# Plot
plt.figure(figsize=(12, 7))
sns.scatterplot(data=df_wealth, x='Exports', y='GDP_per_Capita', hue='Country', s=200, palette='deep')
plt.yscale('log')
plt.title('Trade Openness (2024) vs. National Wealth (Log Scale)', fontsize=14, fontweight='bold')
plt.grid(True, which="both", ls="--", alpha=0.5)
plt.show()
```

### **ANALYSIS AND CONCLUSIONS (OBJECTIVE 10)**
**Objective 10: Trade Openness (2024) vs. National Wealth (Log Scale)**

**1. Key observations from the chart**
* **The High-Income Cluster (G7):**
    * There is a distinct cluster at the top of the chart representing the wealthiest nations (**USA, Germany, UK, France, Japan**), all maintaining a GDP per capita above **$30,000** (log scale).
    * **The United States** is the primary outlier in this group: it has the highest wealth (near **$10^5$**) but the lowest trade openness (Exports ~11%), suggesting its wealth is driven by a massive internal economy.
* **The "Export-Led" Wealth Model (Germany):**
    * **Germany** sits at the far right of the chart, showing the highest trade openness (over **40%**) while maintaining a very high GDP per capita. This confirms that Germany successfully converted its export strategy into national prosperity.
* **The Emerging Market Gap:**
    * **India and China** occupy the lower section of the chart. While their trade openness (Exports ~20%) is similar to Japan's, their GDP per capita remains significantly lower (between **$10^3$ and $1.5 \times$ $10^4$**).
    * **India** is currently at the bottom of this group, highlighting the vast potential for per-capita wealth growth as it integrates further into global markets.

**2. Trends and insights**
* **Non-Linear Relationship:** The use of a **Log Scale** for GDP per capita reveals that moving from a developing to a developed nation requires more than just increasing exports; it involves moving up the value chain.
* **Internal Market vs. Trade Reliance:** The chart illustrates two different paths to wealth: the **US model** (Internal Market focus) vs. the **German/UK model** (Global Trade focus). Both lead to high national wealth, but through different structural dependencies.
* **The Middle-Income Transition:** **China** is visibly "climbing" toward the G7 cluster, with its position significantly higher than India's, reflecting its successful multi-decade industrial and trade expansion.

**3. Conclusion**
The analysis confirms that while high trade openness (Exports % GDP) often correlates with national wealth in Europe, it is not the only path to prosperity, as evidenced by the United States. For emerging giants like **India and China**, the challenge lies in translating their trade volumes into higher per-capita income to bridge the gap with the G7 nations.
