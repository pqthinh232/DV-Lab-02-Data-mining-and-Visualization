### Load the dataset

```python
df = pd.read_csv("data.csv")
print(df.shape)
df.head()
```

### Identify year columns

`year_cols_orig` contains the value variables (the years). It is dynamically generated by iterating through the DataFrame's columns and selecting those that contain a bracket `[`.

```python
year_cols_orig = [col for col in df.columns if '[' in col]
```

```python

print(year_cols_orig)
df[year_cols_orig].head()
```

### Handle non-standard missing values ('..' to NaN) and convert to float

```python
df[year_cols_orig] = df[year_cols_orig].replace('..', np.nan).astype(float)
```

Replace('..', np.nan): Finds all instances of '..' (which is the World Bank's specific placeholder for missing data) and replaces them with np.nan

### Clean column names to just the year integers for accurate time-series 
E.g., 2000 [YR2000] -> 2000

```python
year_mapping = {col: int(col.split(' ')[0]) for col in year_cols_orig}
df = df.rename(columns=year_mapping)
```

```python
print("Dataset Shape:", df.shape)
print("Time Horizon:", min(year_mapping.values()), "to", max(year_mapping.values()))
```

```python
df.head()
```

## 1.3 Handle Missing Values

### Missing Values Analysis

```python
df_ts = df[year_mapping.values()]

missing_per_year = df_ts.isna().sum()
missing_pct_per_year = (missing_per_year / len(df)) * 100
```

Extract only the time series portion (the year columns) for analysis and compute missing value count and percentage per Year.

```python
missing_summary = pd.DataFrame({
    'Missing_Count': missing_per_year,
    'Missing_Percentage (%)': missing_pct_per_year
})

print("\n--- Missing Values Per Year ---")
print(missing_summary)
```

Visualize missing patterns using a Seaborn heatmap.

```python
plt.figure(figsize=(14, 8))

# Dark Navy Blue for "Present" (False/0) and Bright Yellow for "Missing" (True/1)
custom_cmap = sns.color_palette(["#112A46", "#FCE029"])

# The y-axis is the row index (Country/Series), x-axis is the Year
sns.heatmap(df_ts.isna(), cbar=False, cmap=custom_cmap)

# Enhanced typography for the labels
plt.title('Missing Values Heatmap (Yellow = Missing, Dark Blue = Present)', fontsize=16, fontweight='bold', pad=15)
plt.ylabel('Row Index (Unique Country + Series)', fontsize=12)
plt.xlabel('Year Columns', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate x-labels slightly for better readability
plt.tight_layout()
plt.show()
```

Analysis of Missing Patterns:
- Horizontal gaps (Solid yellow lines stretching left to right) indicate specific Country/Series metrics are completely untracked.
- Vertical gaps (like solid yellow bands on specific years) indicate global dataset-wide missing years.
- Trailing missing values on the far right (2022-2024) show data that hasn't been collected or reported yet.

### Preprocessing Data

Convert all data in time-series table to float type

```python
df_ts = df_ts.astype(float)
```

Use a combination of 3 steps:
- Fill in the middle (interpolate).
- Backward fill for the beginning.
- Forward fill for the end.

```python
df_calc = df_ts.interpolate(method='values', axis=1) # Fill in the middle (interpolate)
df_calc = df_calc.bfill(axis=1)                      # Backward fill for the beginning
df_calc = df_calc.ffill(axis=1)                      # Forward fill for the end 
```

Apply to the original table to preserve data: Fill the missing values in the original table using the calculated table above .

```python
id_vars = ["Country Name", "Country Code", "Series Name", "Series Code"]
```

```python
df_time = df_ts.fillna(df_calc).astype(float)
df_final = pd.concat([df[id_vars], df_time], axis=1)
```

Analysis missing value after imputing.

```python
df_ts_final = df_final[year_mapping.values()]

missing_per_year = df_ts_final.isna().sum()
missing_pct_per_year = (missing_per_year / len(df)) * 100
```

```python
missing_summary = pd.DataFrame({
    'Missing_Count': missing_per_year,
    'Missing_Percentage (%)': missing_pct_per_year
})

print("\n--- Missing Values Per Year ---")
print(missing_summary)
```

# Electricity Generation Analysis (Objective 1)

#### **1. Indicators Used**

The analysis utilizes World Bank Development Indicators focusing on the percentage of total electricity production from various sources:

- **Coal:** EG.ELC.COAL.ZS  
- **Natural Gas:** EG.ELC.NGAS.ZS  
- **Oil:** EG.ELC.PETR.ZS  
- **Renewables (Excluding Hydro):** EG.ELC.RNWX.ZS  
- **Hydroelectric:** EG.ELC.HYRO.ZS  
- **Nuclear:** EG.ELC.NUCL.ZS  
- **Other:** A calculated field representing the remaining percentage to ensure the total equals 100%.

#### **2. Reason for Choosing These Indicators**

These indicators were selected to provide a comprehensive view of the global energy transition. By tracking the share of different fuel sources across major economies (`USA`, `China`, `Japan`, `Germany`, `India`, `UK`, and `France`).

Using **percentage of total generation** instead of absolute values enables direct comparison across countries, regardless of differences in grid size.

#### **3. Chart Type**

**Line Chart with Markers**

#### **4. Reason for Choosing This Chart**

The Line Chart with Markers was selected for the following reasons:

* **Temporal Trends**
Line charts are ideal for visualizing time-series data, making it easy to identify trends over a 24-year period.

* **Intersection Points**
Markers highlight individual yearly data points, helping to pinpoint key moments.

* **Clarity of Comparison**
Plotting multiple lines on the same axis allows:
- Easy comparison between energy sources.
- Identification of volatility.
- Recognition of stable sources.

* **Continuity**
The connecting lines emphasize: The rate of change over time.  

```python
# ---------------- 1. CONFIGURATION ----------------
codes = [
    'EG.ELC.COAL.ZS', 'EG.ELC.NGAS.ZS', 'EG.ELC.PETR.ZS',
    'EG.ELC.RNWX.ZS', 'EG.ELC.HYRO.ZS', 'EG.ELC.NUCL.ZS'
]
target_countries = ['United States', 'China', 'Japan', 'Germany', 'India', 'United Kingdom', 'France']
rename_map = {
    'Electricity production from coal sources (% of total)': 'Coal',
    'Electricity production from natural gas sources (% of total)': 'Natural Gas',
    'Electricity production from oil sources (% of total)': 'Oil',
    'Electricity production from renewable sources, excluding hydroelectric (% of total)': 'Renewables',
    'Electricity production from hydroelectric sources (% of total)': 'Hydro',
    'Electricity production from nuclear sources (% of total)': 'Nuclear'
}
years = list(range(2000, 2025))
color_map = {
    'Renewables': '#2ECC71', 
    'Hydro': '#3498DB',      
    'Nuclear': '#9B59B6',    
    'Natural Gas': '#F1C40F',
    'Oil': '#E74C3C',        
    'Coal': '#0B3D91',     
    'Other': '#BDC3C7' 
}

# ---------------- 2. DATA PROCESSING ----------------
# Filter and rename
df_energy = df_final[
    df_final['Series Code'].isin(codes) &
    df_final['Country Name'].isin(target_countries)
].copy()
df_energy['Series Name'] = df_energy['Series Name'].replace(rename_map)

# Calculate "Other" (100% - sum of known sources)
sum_sources = df_energy.groupby('Country Name')[years].sum(min_count=1)
df_other = (100 - sum_sources).clip(lower=0)
df_other['Series Name'] = 'Other'
df_other = df_other.reset_index() 

# Combine and calculate average across all target countries
df_combined = pd.concat([df_energy, df_other], ignore_index=True)
df_avg = df_combined.groupby('Series Name')[years].mean()

# ---------------- 3. PLOTTING ----------------
fig, ax = plt.subplots(figsize=(12, 6))

# Dynamically generate the correct color list based on the dataframe's current index (Series Names).
# Using .get() ensures that if a series name is missing from the map, it defaults to gray instead of crashing.
plot_colors = [color_map.get(col, '#999999') for col in df_avg.index]

df_avg.T.plot(
    ax=ax,
    kind='line',
    marker='o',
    markersize=3,
    linewidth=2,
    color=plot_colors # Apply the dynamically mapped colors
)

# Formatting
ax.set_title(f"Average Energy Trends ({years[0]}-{years[-1]})", fontsize=14, fontweight='bold')
ax.set_ylabel("Percentage (%)")
ax.set_xlabel("Year")

ax.grid(True, alpha=0.3)
ax.legend(title="Energy Source", fontsize=9, loc='upper left', bbox_to_anchor=(1.02, 1))

plt.tight_layout()
plt.show()
```

### **Analysis and Conclusion (Objective 1)**

#### **1. Key Observations from the Chart**

Based on the visualized averages for selected major economies (United States, China, Japan, Germany, India, the United Kingdom, and France), several key patterns emerge:

* **Coal's Dominance and Decline**
Coal remains the largest source of electricity generation at the beginning of the period, accounting for roughly **45% in 2000**.  
  - It peaks around **2012–2013**  
  - Then declines sharply to approximately **30% by 2024**

* **Rapid Rise of Renewables**
Renewable energy (excluding hydro) exhibits the most dramatic growth:
  - Starts at just **1–2% in 2000**  
  - Accelerates rapidly after **2009**  
  - Reaches nearly **20% by 2024**

* **Steady Decline of Nuclear**
Nuclear energy shows a consistent downward trend:
  - Falls from about **26% in 2000**  
  - Drops to roughly **16% by 2024**

* **Growth and Stabilization of Natural Gas**
Natural gas increases steadily over time:
  - Rises from around **14% in 2000**  
  - Peaks slightly above **20% around 2020**  
  - Stabilizes just below **20% in recent years**

* **Stability in Hydroelectric Power**
Hydropower remains the most stable energy source:
  - Maintains a consistent share between **8% and 10%**  
  - Shows minimal fluctuation over the entire period

* **Phasing Out of Oil**
Oil-based electricity generation steadily declines:
  - Starts at about **4% in 2000**  
  - Falls to nearly **1% by 2024**

#### **2. Trends and Insights**

* **The Energy Transition is Clearly Visible**
The data strongly reflects the global shift toward cleaner energy:
  - A clear inverse relationship exists between **Coal** and **Renewables**  
  - As coal declines sharply after **2013**, renewables rise rapidly to fill the gap  

* **Natural Gas as a "Bridge Fuel"**
Natural gas plays a transitional role:
  - Its growth coincides with coal’s decline  
  - Provides a **lower-carbon alternative** while renewable capacity scales up  
  - Helps maintain grid reliability during the transition  

* **Shifting Baseload Power Structure**
Traditionally, **coal** and **nuclear** dominated baseload generation. However:
  - Both show clear decline across these economies  
  - By **2024**, the energy mix becomes more balanced  
  - **Natural Gas, Renewables, and Nuclear** converge in the **16–20% range**

* **Important Intersection Points**
A key milestone occurs around **2020**:
  - **Renewables surpass Nuclear**  
  - Marks a symbolic and practical shift toward modern clean energy technologies  


#### **3. Conclusion**

The period from **2000 to 2024** represents a transformative era in electricity generation for major global economies.

Key takeaways include:
  - The decline of coal’s long-standing dominance  
  - The rapid expansion of renewable energy  
  - The transitional role of natural gas  
  - The stagnation or decline of traditional sources like nuclear and oil  

Although fossil fuels (**coal and natural gas**) still contribute roughly **half of the total energy mix**, current trends indicate a clear movement toward:

- **Decarbonization**  
- **Diversification of energy sources**  
- **Greater reliance on sustainable technologies**

Overall, the data highlights a decisive shift toward a cleaner and more resilient global energy system.

# Cross-Country Electricity Mix Comparison (Objective 2)

#### **1. Indicators Used**

This analysis examines the **percentage of total electricity production** by energy source for the year **2024**. The categories include:

- **Coal:** EG.ELC.COAL.ZS  
- **Natural Gas:** EG.ELC.NGAS.ZS  
- **Oil:** EG.ELC.PETR.ZS  
- **Renewables (Excluding Hydro):** EG.ELC.RNWX.ZS  
- **Hydroelectric:** EG.ELC.HYRO.ZS  
- **Nuclear:** EG.ELC.NUCL.ZS  
- **Other:** A calculated field representing the remaining percentage to ensure the total equals 100%.

#### **2. Reason for Choosing This Indicator**

Using **percentage of total generation** for a single year (2024) enables a clear **cross-sectional comparison** of national energy mixes. This approach makes it possible to directly compare large producers like China with smaller systems like the United Kingdom.

#### **3. Chart Type**

**100% Stacked Bar Chart**

#### **4. Reason for Choosing This Chart**

The **100% Stacked Bar Chart** is ideal for visualizing **part-to-whole relationships** across multiple countries.

* Side-by-Side Comparison
  - Each country is represented as a single bar  
  - Enables immediate comparison of energy mixes across nations  

* Visual Proportions
  - Color-coded segments clearly show the share of each energy source  
  - Makes dominant sources instantly recognizable.

* Standardized Baseline
  - All bars are normalized to **100% height**  
  - Allows easy estimation of relative contributions within each country  
  - Improves readability and interpretability of proportional data  

Overall, this visualization provides a concise and intuitive way to compare how different economies structure their electricity generation in 2024.

```python
# ---------------- 1. DATA PREPARATION ----------------
year_to_plot = 2024

# Extract just the Country, Series Name, and the 2024 values
df_2024 = df_combined[['Country Name', 'Series Name', year_to_plot]].copy()
df_pivot = df_2024.pivot(index='Country Name', columns='Series Name', values=year_to_plot)

# ---------------- 2. ORDER ----------------
ordered_list = ['Renewables', 'Hydro', 'Nuclear', 'Natural Gas', 'Oil', 'Coal', 'Other']

# Reorder the columns to match the legend hierarchy in the image
ordered_cols = [col for col in ordered_list if col in df_pivot.columns]
df_pivot = df_pivot[ordered_cols]

# Create the final color list in the same exact order
colors = [color_map[col] for col in ordered_cols]

# ---------------- 3. PLOTTING ----------------
fig, ax = plt.subplots(figsize=(14, 8))

# Plotting the stacked bar chart directly from the pivoted dataframe
df_pivot.plot(
    kind='bar', 
    stacked=True, 
    color=colors, 
    ax=ax,
    width=0.75
)

# Formatting the chart to match the image
ax.set_title(f"3. Comparison of Electricity Mix across Countries ({year_to_plot})", fontsize=16, fontweight='bold')
ax.set_ylabel("Percentage (%)", fontsize=12)
ax.set_xlabel("") 
ax.set_ylim(0, 102) 

plt.xticks(rotation=0)
ax.legend(bbox_to_anchor=(1.015, 1), loc='upper left', borderaxespad=0.)

plt.tight_layout()
plt.show()
```

# Cross-Country Population Structure (Objective 3)

#### **1. Indicators Used**

This analysis examines demographic distribution for the year **2024**, focusing on the percentage split of each country’s total population:

- **Urban Population:** Percentage of people living in urban areas  
- **Rural Population:** Percentage of people living in rural areas  

#### **2. Reason for Choosing This Indicator**

The **urban–rural divide** provides a clear snapshot of a country’s:

- Level of **urbanization**  
- Stage of **economic development**  
- **Labor structure** (industrial vs. agricultural)  
- **Infrastructure demand**  

This indicator highlights the contrast between:
- Highly developed, industrialized nations (typically **urban-dominated**)  
- Developing economies with significant **rural populations**  

#### **3. Chart Type**

**Horizontal Stacked Bar Chart**

#### **4. Reason for Choosing This Chart**

This visualization was selected for its effectiveness in comparing proportional data across countries:

* **Clear Part-to-Whole Representation**
  - Each bar represents **100% of a country’s population**  
  - Clearly shows the split between **urban** and **rural** populations  

* **Fair Cross-Country Comparison**
  - Normalization removes bias from population size differences  
  - Enables direct comparison between countries like:
    - **India** (very large population)  
    - **France** (much smaller population)  

* **Improved Readability**
  - Horizontal layout keeps **country names fully readable**  
  - Avoids clutter and overlapping labels  

* **Immediate Visual Contrast** Distinct color segments (e.g., teal vs. orange) make patterns obvious:

Overall, this chart provides an intuitive and visually clean way to compare **population structure and urbanization levels** across different countries in 2022.

```python
# --- 1. THEME & SETUP ---
sns.set_theme(style="whitegrid")
plt.rcParams.update({
    'font.family': 'sans-serif',
    'axes.titlesize': 13,
    'axes.labelsize': 11,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'legend.fontsize': 10
})

countries = ['United States', 'China', 'Japan', 'Germany', 'India', 'United Kingdom', 'France']
series_urban = 'Urban population (% of total population)'

year_cols = [col for col in df.columns if str(col).isdigit()]

target_year = 2024

# --- 2. DATA PREPARATION ---
df_u_target = df[(df['Series Name'] == series_urban) & (df['Country Name'].isin(countries))]
urban_vals = pd.to_numeric(df_u_target.set_index('Country Name')[target_year], errors='coerce')
rural_vals = 100 - urban_vals

# Build plotting dataframe and sort by Urban percentage
plot_df = pd.DataFrame({'Urban': urban_vals, 'Rural': rural_vals}).sort_values('Urban')

# --- 3. PLOTTING ---
fig, ax = plt.subplots(figsize=(10, 6))

color_urban = '#2a9d8f'  # Teal
color_rural = '#e76f51'  # Terracotta

# Plot horizontal stacked bar chart
plot_df.plot(
    kind='barh', 
    stacked=True, 
    ax=ax, 
    color=[color_urban, color_rural], 
    alpha=0.9, 
    width=0.6
)

# Formatting
ax.set_title(f'Population Structure across Countries (Year {target_year})', fontweight='bold', fontsize=14)
ax.set_xlabel('Percentage (%)')
ax.set_ylabel('') 

# Set limits and legend
ax.set_xlim(0, 100)
ax.legend(loc='lower center', bbox_to_anchor=(0.5, -0.2), ncol=2, frameon=False)

plt.tight_layout()
plt.show()
```

# Comparison of Threatened Species by Type (Objective 4)

#### **1. Indicators Used**
This analysis examines biodiversity data for the year **2022**, selected because it is the most recent year with available data. The focus is on the **absolute number of threatened species** within national borders.

The data is categorized into four taxonomic groups:
- Bird
- Fish
- Mammal
- Plant


#### **2. Reason for Choosing This Indicator**
Tracking the number of threatened species by category provides a clear and measurable view of a country’s biodiversity health and ecological stress levels.

* Key Benefits:
  - Identifies specific conservation challenges (e.g., declining fish populations vs. plant biodiversity loss)
  - Moves beyond a single aggregate metric to reveal hidden patterns across species groups
  - Supports targeted environmental policies and conservation strategies

#### **3. Chart Type**
**Grouped Bar Chart**

#### **4. Reason for Choosing This Chart**

The grouped bar chart is particularly effective for comparing multiple sub-categories across countries:

* Side-by-Side Sub-category Comparison
  - Enables direct comparison within a single country
  - Allows quick assessment of whether birds, fish, mammals, or plants are most at risk

* Magnitude and Scale
  - Displays **absolute values** rather than proportions
  - Highlights disparities between:
    - Biodiversity-rich nations (large ecosystems)
    - Smaller or less biodiverse countries

* Clear Categorization
  - Distinct colors represent each species group
  - Clustered layout prevents data overlap
  - Makes the visualization easy to read and interpret quickly

```python
# --- 1. CONFIG ---
target_year = 2022

threatened_series = [
    'Bird species, threatened',
    'Fish species, threatened',
    'Mammal species, threatened',
    'Plant species (higher), threatened'
]

countries = [
    'United States', 'China', 'Japan',
    'Germany', 'India', 'United Kingdom', 'France'
]

# Short names for legend
short_names = {
    'Bird species, threatened': 'Bird',
    'Fish species, threatened': 'Fish',
    'Mammal species, threatened': 'Mammal',
    'Plant species (higher), threatened': 'Plant'
}

# --- 2. DATA PREPARATION ---
pivot_df = (
    df_final
    .loc[
        df_final['Series Name'].isin(threatened_series) &
        df_final['Country Name'].isin(countries),
        ['Country Name', 'Series Name', target_year]
    ]
    .assign(**{'Series Name': lambda x: x['Series Name'].replace(short_names)})
    .pivot(index='Country Name', columns='Series Name', values=target_year)
    [['Bird', 'Fish', 'Mammal', 'Plant']]  # ensure order
)

# --- 3. PLOTTING ---
fig, ax = plt.subplots(figsize=(12, 6))

pivot_df.plot(
    kind='bar',
    ax=ax,
    colormap='viridis', 
    width=0.7
)

# Formatting
ax.set_title(f"Comparison of Threatened Species by Type ({target_year})",
             fontsize=15, fontweight='bold', pad=15)
ax.set_ylabel("Number of Species", fontweight='bold')
ax.set_xlabel("")
ax.tick_params(axis='x', rotation=45)

# Grid & legend
ax.grid(axis='y', linestyle='--', alpha=0.6)
ax.legend(title="Species Type", framealpha=0.9)

plt.tight_layout()
plt.show()
```
