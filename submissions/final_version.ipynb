## 1.2 Load and Preprocess Data

#### **Dataset Overview**
The dataset is built from the **World Development Indicators (WDI)**, a premier compilation of cross-country, comparable statistics collected by the World Bank. It is specifically curated to evaluate and compare the developmental trajectories, economic shifts, societal changes, and environmental footprints of seven major global powers over a quarter-century.

* **Time Horizon:** 25 years (2000 – 2024).
* **Target Entities (7 Countries):** United States, China, Japan, Germany, India, United Kingdom, and France.
* **Core Domains:** The data is divided into three primary analytical pillars:
    * **Economy:** GDP growth, GDP per capita, structural value-added (Agriculture, Industry, Services), export performance, and inflation.
    * **Society:** Tertiary education enrollment, demographic breakdowns (0-14, 15-64, 65+), urbanization rates, internet adoption, life expectancy, infant mortality, healthcare expenditure, unemployment, and gender participation in the labor force.
    * **Environment:** Electricity generation mix (Coal, Gas, Oil, Hydro, Nuclear, Renewables) and biodiversity threats (threatened birds, mammals, fish, and plants).


#### **Data Structure and Dimensions**
The raw dataset is structured in a **"wide" panel data format**, where each row represents a unique combination of a country and a specific indicator, and the columns represent the years.

* **Number of Unique Indicators:** 33 distinct macroeconomic, social, and environmental metrics.
* **Number of Records (Rows):** 231 records (Calculated as $7 \text{ countries} \times 33 \text{ unique indicators}$).
* **Number of Features (Columns):** 29 columns (Calculated as 4 identifier columns + 25 year columns).

#### Feature Breakdown
1.  **Identifier Features (4):** `Country Name`, `Country Code`, `Series Name`, `Series Code`.
2.  **Time-Series Features (25):** Columns representing each year from 2000 to 2024 (originally formatted as `YYYY [YRYYYY]` before cleaning).

#### **Data Quality & Preprocessing Characteristics**
* **Missing Values Representation:** The World Bank's standard placeholder for missing data (`..`) is present in the raw dataset.
* **Data Completeness:** * Certain variables (e.g., GDP growth for specific nations) feature complete, unbroken 25-year series.
    * Other variables (e.g., tertiary education enrollment for the US, Japan, and Germany) contain historical gaps or trailing missing values for the most recent years (2023-2024) due to reporting lags.

#### Imputation Strategy
To ensure continuous trend analysis without data leakage or misleading drops, a three-step algorithmic imputation was applied to the time-series grid:

1.  **Linear Interpolation:** To bridge gaps between known data points.
2.  **Backward Fill (bfill):** To estimate missing initial years based on the earliest recorded baseline.
3.  **Forward Fill (ffill):** To carry the most recent known values forward into unreported trailing years (like 2023 or 2024).

Convert all data in time-series table to float type

```python
df_ts = df_ts.astype(float)
```

Use a combination of 3 steps:
- Fill in the middle (interpolate).
- Backward fill for the beginning.
- Forward fill for the end.

```python
df_calc = df_ts.interpolate(method='values', axis=1) # Fill in the middle (interpolate)
df_calc = df_calc.bfill(axis=1)                      # Backward fill for the beginning
df_calc = df_calc.ffill(axis=1)                      # Forward fill for the end 
```

Apply to the original table to preserve data: Fill the missing values in the original table using the calculated table above .

```python
id_vars = ["Country Name", "Country Code", "Series Name", "Series Code"]
```

```python
df_time = df_ts.fillna(df_calc).astype(float)
df_final = pd.concat([df[id_vars], df_time], axis=1)
```

## Analysis Objective 1: Identifying Growth Leaders and Assessing the Impact of Global Crises (2000 - 2024)

### 1. Indicator Used
* **Data field:** `GDP growth (annual %)`
* **Meaning of the indicator:** This indicator measures the annual percentage growth rate of GDP at constant prices using local currency. It reflects the immediate "health" of an economy, showing whether the economy is expanding or contracting compared to the previous year.

### 2. Reason for Choosing This Indicator
* **Relevance to the objective:** To identify "growth leaders," the annual GDP growth rate is the fairest metric for comparing economies of different sizes. For example, it clearly highlights India’s rapid breakthroughs as well as the more stable performance of larger economies like the United States.
* **Why this indicator:** GDP growth is highly sensitive to external shocks. Tracking this indicator allows us to clearly see the economic "scars" left by major global crises such as the 2008 Global Financial Crisis and the COVID-19 pandemic on each country.

### 3. Chart Type: **Line Chart with Markers and Reference Lines**

### 4. Reason for Choosing This Chart
* **Nature of the data:** The data fluctuates significantly from year to year. A line chart is ideal for visualizing the rhythm of economic cycles, including peaks and troughs over time.

* **Handling real-world data:**
    * **Markers (Points):** These help pinpoint the exact growth rate for each specific year. This is particularly important when comparing crisis years, where growth can drop into negative territory.
    * **Zero-line reference:** A horizontal reference line at $y = 0$ is essential to clearly distinguish between positive growth (even if slow) and actual recession (negative growth).

* **Comparison ability:** The chart makes it easy to observe the contrast between emerging economies and developed economies.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load raw data 
df_raw = pd.read_csv('../dataset/data.csv') 
df_raw = df_raw.replace(' .. ', np.nan)

gdp_growth_indicator = "GDP growth (annual %)"
df_gdp_base = df_raw[df_raw['Series Name'] == gdp_growth_indicator].copy()
year_cols = [col for col in df_gdp_base.columns if '[' in col]

# Convert data to numeric (float)
df_gdp_base[year_cols] = df_gdp_base[year_cols].apply(pd.to_numeric, errors='coerce')

# Data preparation
# Dataset: Convert to Long format for plotting with Seaborn
df_plot_long = df_gdp_base.melt(id_vars=['Country Name'], value_vars=year_cols, 
                                 var_name='Year_Raw', value_name='Growth_Rate')
df_plot_long['Year'] = df_plot_long['Year_Raw'].str.extract(r'(\d+)')[0].astype(int)

target_countries = [
    'United States', 'China', 'Japan',
    'Germany', 'India', 'United Kingdom', 'France'
]
df_filtered = df_plot_long[df_plot_long['Country Name'].isin(target_countries)]

# plotting
plt.figure(figsize=(15, 8))
sns.set_style("whitegrid")

# Create fixed color palette
palette = sns.color_palette("bright", n_colors=len(target_countries))

# Draw trend lines
sns.lineplot(
    data=df_filtered, 
    x='Year', y='Growth_Rate', hue='Country Name', 
    palette=palette, linewidth=2, alpha=0.8, marker='o', markersize=6
)

# Add reference line y=0
plt.axhline(0, color='black', linestyle='--', linewidth=1.5, alpha=0.7)

# Mark crisis periods
plt.axvspan(2008, 2009, color='gray', alpha=0.2, label='2008 Financial Crisis')
plt.axvspan(2020, 2021, color='red', alpha=0.1, label='COVID-19 Pandemic')

plt.title('GDP Annual Growth Rate: Leaders vs. Global Crises (2000 - 2024)', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP Growth (Annual %)', fontsize=12)
plt.legend(title='Country', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.xticks(range(2000, 2025, 2), fontsize=10)
plt.xlim(2000, 2024)

plt.tight_layout()
plt.show()
```

## Analysis and Conclusions (GDP Growth Trends)

### 1. Key Observations from the Chart

## Analysis Objective 2: Measuring Economic Convergence and Wealth Inequality through GDP per Capita (2000–2024)

### 1. Indicator Used
* **Data field:** `GDP per capita (current US$)`
* **Meaning of the indicator:** This indicator is calculated by dividing a country’s total GDP by its population in a given year. It shows the average income level per person and is one of the most common ways to measure and compare living standards across countries, even when their population sizes are very different.

### 2. Reason for Choosing This Indicator
* **Relevance to the objective:** While total GDP tells us how big an economy is, GDP per capita helps us understand the “quality” of that growth — how wealthy people actually are on average. This indicator allows us to examine whether developing countries are really catching up with richer nations in terms of income and living standards.
* **Why this indicator:** It clearly highlights the large income gaps between countries (for example, between the United States and India). These gaps reflect differences in labor productivity, technology levels, and overall wealth inequality across the world.

### 3. Chart Type: **Line Chart**

### 4. Reason for Choosing This Chart
* **Nature of the data:** GDP per capita varies greatly (from $2,695 to $84,534). 
* **Handling real-world data:**
    * **Linear Scale:** By using a standard linear scale, the chart vividly illustrates the **absolute wealth gap**. It emphasizes how massive the distance is between an average citizen in the US versus one in India or China.
    * **Highlighting Gaps:** The vertical distance between lines represents the real-world inequality in purchasing power and living standards.

```python
# 1. Data preparation
gdp_pc_indicator = "GDP per capita (current US$)"
# Filter the correct indicator from the list
df_pc_base = df_raw[df_raw['Series Name'] == gdp_pc_indicator].copy()

# Identify year columns based on Index list
year_cols = [col for col in df_pc_base.columns if '[' in col]

df_pc_base[year_cols] = df_pc_base[year_cols].apply(pd.to_numeric, errors='coerce')
# Convert to Long format
df_pc_long = df_pc_base.melt(id_vars=['Country Name'], value_vars=year_cols, 
                             var_name='Year_Raw', value_name='GDP_PC')

# Extract year (e.g., '2000 [YR2000]' -> 2000)
df_pc_long['Year'] = df_pc_long['Year_Raw'].str.extract(r'(\d+)')[0].astype(int)

# List of countries exactly matching your data
target_countries = ['United States', 'China', 'Japan', 'Germany', 'India', 'United Kingdom', 'France']
df_pc_filtered = df_pc_long[df_pc_long['Country Name'].isin(target_countries)]

# 2.Plotting
plt.figure(figsize=(15, 8))
sns.set_style("whitegrid")

# Use a color palette that clearly distinguishes countries
palette = sns.color_palette("husl", n_colors=len(target_countries))

# Draw trend lines
sns.lineplot(
    data=df_pc_filtered, 
    x='Year', y='GDP_PC', hue='Country Name', 
    palette=palette, linewidth=2.5, marker='o', markersize=5, alpha=0.8
)

# 3.
for country in target_countries:
    # Filter data for each country and drop NaN to get the actual last value
    country_data = df_pc_filtered[df_pc_filtered['Country Name'] == country].dropna(subset=['GDP_PC'])
    
    if not country_data.empty:
        # Get the row with the largest year that has data
        last_row = country_data.sort_values('Year').iloc[-1]
        last_val = last_row['GDP_PC']
        last_year = last_row['Year']
        
        # Draw value label at the last data point
        plt.text(last_year + 0.3, last_val, f'${last_val:,.0f}', 
                 fontsize=10, fontweight='bold', va='center', color='black')

plt.title('The Growing Wealth Gap: GDP Per Capita Trends (2000 - 2024)', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP per capita (current US$)', fontsize=12)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(range(2000, 2025, 2))
plt.xlim(2000, 2026) # Expand right margin so labels aren't cut off

plt.tight_layout()
plt.show()
```

### 2. Reason for Choosing This Indicator
* **Relevance to the objective:** To see which countries are experiencing deindustrialization and how fast it is happening, we need to track the changing share of industry in GDP over time.
* **Why this indicator:** A decreasing percentage usually indicates deindustrialization. This process is often linked to:

    * Moving manufacturing activities to other countries (offshoring)
    * Broader structural changes in the economy

### 3. Chart Type: **Line Chart with Markers**

### 4. Reason for Choosing This Chart

## Analysis Objective 5: Service Sector Dominance and Structural Maturity

### 1. Indicator Used

### 4. Reason for Choosing This Chart
* **Nature of the data:** A **Boxenplot** is superior for showing the distribution, median, and outliers across different countries.
* **Clarity:** A supplementary **Lineplot** is used to visualize the specific trajectory of growth for each nation, making it easy to spot the "strongest increase."

```python
#Load data
df_imputed = pd.read_csv("../dataset/data_imputed.csv")

# Filter indicator
service_indicator = "Services, value added (% of GDP)"
df_services = df_imputed[df_imputed['Series Name'] == service_indicator].copy()

# Melt to long format for Seaborn
year_cols = [str(y) for y in range(2000, 2025)]
df_services_long = df_services.melt(id_vars=['Country Name'], value_vars=year_cols, 
                                    var_name='Year', value_name='Services_pct')

import matplotlib.pyplot as plt
import seaborn as sns

# 1. Tính toán thứ tự các nước theo trung bình % Dịch vụ để tạo sự liên kết
order = df_services_long.groupby("Country Name")["Services_pct"].median().sort_values(ascending=False).index

# 2. Thiết lập bảng màu đồng nhất cho tất cả các nước dựa trên thứ tự này
palette = sns.color_palette("viridis", len(order))
color_map = dict(zip(order, palette))

# 3. Tạo Figure
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 12))

# --- BIỂU ĐỒ 1: BOXENPLOT (Dùng order và color_map) ---
sns.boxenplot(data=df_services_long, x='Country Name', y='Services_pct', 
              order=order, palette=color_map, ax=ax1)
ax1.set_title('1. Structural Maturity: Distribution of Services % GDP (Ordered by Median)', fontsize=14, fontweight='bold')
ax1.set_ylabel('Services % of GDP')
ax1.tick_params(axis='x', rotation=30)

# --- BIỂU ĐỒ 2: LINEPLOT (Dùng color_map để cùng màu với biểu đồ trên) ---
sns.lineplot(data=df_services_long, x='Year', y='Services_pct', 
             hue='Country Name', hue_order=order, palette=color_map, 
             marker='o', linewidth=2, ax=ax2)
ax2.set_title('2. Growth Trajectory: Evolution of Service Sectors (2000-2024)', fontsize=14, fontweight='bold')
ax2.set_ylabel('Services % of GDP')
ax2.legend(title='Country (High to Low)', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

## Analysis Objective 8: Structural Transformation of the Economy (2000 vs 2024)

## Analysis Objective 9: The Correlation Between Industrial Strength and GDP Growth

* **Why this indicator:** To investigate if the "industrial engine" model still holds true for these nations in the modern era.

### 3. Chart Type: **Regression Plot (Regplot)**

### 4. Reason for Choosing This Chart
* **Nature of the data:** A **Regplot** combines a scatter plot with a trend line, making the relationship between two continuous variables visible.
* **Clarity:** Including the Pearson correlation coefficient (r) provides a mathematical basis for the conclusion.

```python
!pip install spicy
```

```python
from scipy import stats

# Extract and align two indicators
df_growth = df_imputed[df_imputed['Series Name'] == "GDP growth (annual %)"].set_index('Country Name')[year_cols].stack().reset_index()
df_ind = df_imputed[df_imputed['Series Name'] == "Industry (including construction), value added (% of GDP)"].set_index('Country Name')[year_cols].stack().reset_index()

# Combine
df_corr = pd.DataFrame({
    'Growth': df_growth[0].values,
    'Industry': df_ind[0].values
})

# Calculate Pearson Correlation
r, p_value = stats.pearsonr(df_corr['Industry'], df_corr['Growth'])

# Plot
plt.figure(figsize=(10, 6))
sns.regplot(data=df_corr, x='Industry', y='Growth', 
            scatter_kws={'alpha':0.4, 'color':'#3498db'}, 
            line_kws={'color':'red', 'label': f'Linear Fit (r={r:.2f})'})

# Add Correlation Coefficient text onto the plot
plt.text(16, 13, f'Pearson r = {r:.2f}\np-value < 0.001', 
         fontsize=12, fontweight='bold', color='red', 
         bbox=dict(facecolor='white', alpha=0.5))

plt.title('Correlation: Industrial Contribution vs. Annual GDP Growth', fontsize=14, fontweight='bold')
plt.xlabel('Industry (% of GDP)')
plt.ylabel('Annual GDP Growth (%)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### **ANALYSIS AND CONCLUSIONS (OBJECTIVE 9)**
**Objective 9: Correlation between Industrial Contribution and Annual GDP Growth**

**1. Key observations from the chart**
* **Significant Positive Correlation:**
    * The Pearson correlation coefficient (**r = 0.59**) indicates a **moderate-to-strong positive relationship** between industrial value-added and annual GDP growth.
    * The p-value (**< 0.001**) confirms that this relationship is statistically highly significant, meaning the observed trend is very unlikely to be due to random chance.
* **The "Industrial Sweet Spot" for High Growth:**
    * Data points in the high-industry zone (**Industry > 40%**) consistently coincide with exceptional GDP growth rates exceeding **8% to 12%**. This cluster likely represents periods of rapid industrialization.
* **Stability vs. Volatility:**
    * Economies with an industrial share between **15% and 25%** (typically G7 nations) show a dense cluster of stable growth between **0% and 4%**.
    * However, as the industrial share moves into the **25% - 32%** range, volatility increases significantly, with growth rates ranging from **-6% to nearly 10%**.
* **Vulnerability of Low-Industry Zones:**
    * The most severe negative growth outliers (dipping to -10%) are concentrated in the 15-20% Industry range. This indicates that service-heavy economies in this dataset are more susceptible to deep recessionary shocks compared to those with a larger industrial base.

**2. Trends and insights**
* **Industrial Sector as a Growth Engine:** The upward slope of the red regression line proves that for these 7 major economies, a larger industrial sector acts as a powerful driver of economic expansion.

### 4. Reason for Choosing This Chart
* **Nature of the data:** This is time-series data. A line chart is the most suitable way to show trends, growth rates, and key turning points over the 24-year period.

* **Handling real-world data:**
    * **Transparency:** Some countries (such as the United States, Japan, and Germany) have missing data in earlier years. The group keeps these gaps and only starts plotting from available data. Using **markers (dots)** helps clearly show actual data points and distinguish them from interpolated segments.
    * **Linear interpolation:** For short gaps within the timeline, linear interpolation is used to maintain a smooth trend instead of leaving flat lines or sudden jumps.

* **Comparison ability:** With 7 countries, using different colored lines makes it easier to compare their positions (who leads, who follows) and observe changes through the slope of each line.

```python
# Load raw data
df_raw = pd.read_csv('../dataset/data.csv', na_values='..')
edu_indicator = "School enrollment, tertiary (% gross)"
df_edu_base = df_raw[df_raw['Series Name'] == edu_indicator].copy()
year_cols = [col for col in df_edu_base.columns if '[' in col]

# Interpolated data (used to draw continuous lines)
df_interp = df_edu_base.copy()
df_interp[year_cols] = df_interp[year_cols].interpolate(method='linear', axis=1, limit_area='inside')
df_interp_long = df_interp.melt(id_vars=['Country Name'], value_vars=year_cols, var_name='Year_Raw', value_name='Val')
df_interp_long['Year'] = df_interp_long['Year_Raw'].str.extract(r'(\d+)')[0].astype(int)

# Original data (used for markers - only real values, missing ones remain NaN)
df_dots_long = df_edu_base.melt(id_vars=['Country Name'], value_vars=year_cols, var_name='Year_Raw', value_name='Val')
df_dots_long['Year'] = df_dots_long['Year_Raw'].str.extract(r'(\d+)')[0].astype(int)


plt.figure(figsize=(15, 8))
sns.set_style("whitegrid")

# Use a consistent color palette so both layers match
colors = sns.color_palette("tab10", n_colors=len(df_edu_base['Country Name'].unique()))
palette = dict(zip(df_edu_base['Country Name'].unique(), colors))

# LAYER 1: Draw LINES (using interpolated data, no markers)
sns.lineplot(
    data=df_interp_long, 
    x='Year', y='Val', hue='Country Name', 
    palette=palette, linewidth=2.5, alpha=0.6,
    marker=None, legend=True  # Keep legend here
)

# LAYER 2: Draw POINTS (using original data, missing years will have no points)
sns.scatterplot(
    data=df_dots_long, 
    x='Year', y='Val', hue='Country Name', 
    palette=palette, s=60, edgecolor='white',
    legend=False  # Disable legend here to avoid duplication
)

plt.title('Trends in Tertiary Education Enrollment of Major Countries (2000 - 2024)', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Tertiary Enrollment (% Gross)', fontsize=12)
plt.legend(title='Country', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.xticks(range(2000, 2025, 2), fontsize=10)
plt.xlim(2000, 2024)

plt.tight_layout()
plt.show()
```

#### **3. Chart Type**

#### **4. Reason for Choosing This Chart**

The Line Chart with Markers was selected for the following reasons:

Line charts are ideal for visualizing time-series data, making it easy to identify trends over a 24-year period.

```python
# ---------------- 1. CONFIGURATION ----------------
codes = [
    'EG.ELC.COAL.ZS', 'EG.ELC.NGAS.ZS', 'EG.ELC.PETR.ZS',
    'EG.ELC.RNWX.ZS', 'EG.ELC.HYRO.ZS', 'EG.ELC.NUCL.ZS'
]
target_countries = ['United States', 'China', 'Japan', 'Germany', 'India', 'United Kingdom', 'France']
rename_map = {
    'Electricity production from coal sources (% of total)': 'Coal',
    'Electricity production from natural gas sources (% of total)': 'Natural Gas',
    'Electricity production from oil sources (% of total)': 'Oil',
    'Electricity production from renewable sources, excluding hydroelectric (% of total)': 'Renewables',
    'Electricity production from hydroelectric sources (% of total)': 'Hydro',
    'Electricity production from nuclear sources (% of total)': 'Nuclear'
}
years = list(range(2000, 2025))
color_map = {
    'Renewables': '#2ECC71', 
    'Hydro': '#3498DB',      
    'Nuclear': '#9B59B6',    
    'Natural Gas': '#F1C40F',
    'Oil': '#E74C3C',        
    'Coal': '#0B3D91',     
    'Other': '#BDC3C7' 
}

# ---------------- 2. DATA PROCESSING ----------------
# Filter and rename
df_energy = df_final[
    df_final['Series Code'].isin(codes) &
    df_final['Country Name'].isin(target_countries)
].copy()
df_energy['Series Name'] = df_energy['Series Name'].replace(rename_map)

# Calculate "Other" (100% - sum of known sources)
sum_sources = df_energy.groupby('Country Name')[years].sum(min_count=1)
df_other = (100 - sum_sources).clip(lower=0)
df_other['Series Name'] = 'Other'
df_other = df_other.reset_index() 

# Combine and calculate average across all target countries
df_combined = pd.concat([df_energy, df_other], ignore_index=True)
df_avg = df_combined.groupby('Series Name')[years].mean()

# ---------------- 3. PLOTTING ----------------
fig, ax = plt.subplots(figsize=(12, 6))

# Dynamically generate the correct color list based on the dataframe's current index (Series Names).
# Using .get() ensures that if a series name is missing from the map, it defaults to gray instead of crashing.
plot_colors = [color_map.get(col, '#999999') for col in df_avg.index]

df_avg.T.plot(
    ax=ax,
    kind='line',
    marker='o',
    markersize=3,
    linewidth=2,
    color=plot_colors # Apply the dynamically mapped colors
)

# Formatting
ax.set_title(f"Average Energy Trends ({years[0]}-{years[-1]})", fontsize=14, fontweight='bold')
ax.set_ylabel("Percentage (%)")
ax.set_xlabel("Year")

ax.grid(True, alpha=0.3)
ax.legend(title="Energy Source", fontsize=9, loc='upper left', bbox_to_anchor=(1.02, 1))

plt.tight_layout()
plt.show()
```

### 1. The Economic Shift: 
The global economic order is undergoing a profound structural transformation. Over the past two decades, China and India have emerged as the primary engines of world GDP growth, demonstrating remarkable resilience during major crises such as the 2008 financial meltdown and the 2020 COVID-19 pandemic. While developed economies like the United States maintain absolute dominance in GDP per capita and service sector maturity, the gap in growth momentum has clearly shifted toward Asia.

China stands out with the fastest structural modernization — rapidly reducing its reliance on industry and agriculture while expanding services to over 57% of GDP. Germany continues to defy the trend of deindustrialization, preserving a strong manufacturing base that supports its export-led model. In contrast, Japan faces stagnation in both GDP per capita and industrial vitality, while India still grapples with high agricultural dependence and the challenge of translating demographic advantage into sustained productivity gains.

Ultimately, the economic shift reveals a new multipolar reality: the United States remains the wealthiest superpower, yet the future drivers of global growth increasingly lie with the rising Asian giants. This transition underscores that economic power in the 21st century will be determined not only by current wealth, but by the ability to adapt, industrialize smartly, and harness structural change.


### 2. Social Convergence and Demographic Divergence
The strength of a superpower in 2024 is increasingly defined by "Knowledge Capital" and demographic stability.
*   **The Knowledge Revolution:** One key finding is that the global “knowledge gap” has largely narrowed. By 2024, China has reached levels comparable to the G7 in higher education enrollment and digital access, with over 92% of its population using the Internet.
*   **Demographic Challenges:** Our analysis highlights a clear demographic divide. Aging populations in Japan, Germany, and France pose long-term economic and social challenges. In contrast, India benefits from a large and young population, but its potential will depend on addressing structural issues, especially the gender gap in workforce participation.
*   **System Efficiency:** The study also shows that a high GDP does not automatically translate into a high quality of life. The model in United States faces efficiency issues in healthcare, where high spending does not lead to the same life expectancy outcomes seen in more public-oriented systems like those in Japan and France.

### 3. The Environmental Toll and the Energy Transition

The figures collectively illustrate the complex interplay between global energy transitions, demographic shifts, and environmental impacts. While the global energy trend from 2000 to 2024 shows a rapidly accelerating shift toward renewable energy, fossil fuels—particularly coal—still dominate the overall mix. This dependence is highly regional, as seen in the 2024 electricity mix, where massive, rapidly developing nations like China and India remain heavily reliant on coal, while countries like France and the UK lean heavily on nuclear and renewables, respectively.

Furthermore, these energy profiles are intrinsically linked to population structures; developed nations (like Japan and the UK) are highly urbanized, whereas India maintains a predominantly rural population. Finally, the high volume of threatened species—specifically plants and fish in large countries like China, India, and the United States—highlights the significant ecological toll that sustained development, heavy fossil fuel reliance, and expanding urbanization continue to extract on global biodiversity.
